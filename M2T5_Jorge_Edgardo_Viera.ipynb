**1- Sumatoria de no negativos**



```python
# codigo correspondiente al fichero funcs.py

def sumatoria_positivos(nums):
    """
    Suma solo los números mayores o iguales a 0 de una lista.
    """
    total = 0
    for n in nums:
        if n >= 0:
            total += n
    return total
```

```python
# codigo correspondiente al fichero test_funcs.py

import pytest
from funcs import sumatoria_positivos #importamos la función a testear

@pytest.mark.parametrize("nums, esperado", [
    ([], 0),                         # Lista vacía
    ([-1, -5, -10], 0),              # Solo negativos
    ([1, 2, 3, 4], 10),              # Solo positivos
    ([-1, 2, -3, 4, 0], 6),          # Mixta (incluyendo cero)
    ([0, 0, 0], 0),                  # Solo ceros
    ([-100, 100, -50, 50], 150),     # Mixta con valores grandes
])
def test_sumatoria_positivos(nums, esperado):
    """Prueba parametrizada para la función sumatoria_positivos con diferentes casos de prueba.
    Se incluyen casos con listas vacías, solo negativos, solo positivos, mezclas y ceros para verificar su correcto funcionamiento.
    """
    assert sumatoria_positivos(nums) == esperado # Verificamos que el resultado de la función sea igual al valor esperado para cada caso de prueba
```

**2-Palíndromo básico**

```python
# codigo de la función es_palindromo del fichero funcs.py

def es_palindromo(s: str) -> bool:
    s = s.lower().replace(' ', '')
    return s == s[::-1]
```

**3-Máximo con control de errores**

```python
# codigo correspondiente a la función maximo_seguro del fichero funcs.py

def maximo_seguro(nums):
    if not nums:
        raise ValueError('lista vacía')
    maximo = nums[0]
    for n in nums[1:]:
        if n > maximo:
            maximo = n
    return maximo
```

```python
# codigo de prueba para la función maximo_seguro del fichero test_funcs.py

import pytest
from funcs import maximo_seguro

@pytest.mark.parametrize("nums, esperado", [
    ([1, 2, 3, 4, 5], 5),          # Lista con números positivos
    ([-1, -2, -3, -4, -5], -1), # Lista con números negativos
    ([5], 5),                    # Lista con un solo elemento
    ([-10, -5, 0, 5, 10], 10),   # Lista mixta con negativos y positivos
    ([1.5, 2.5, 3.5], 3.5),     # Lista con números flotantes
    ([-1.5, -2.5, -0.5], -0.5), # Lista con números flotantes negativos
    #lista vacia descomentar para probar la excepcion
    #([], None)                    # Lista vacía, se espera una excepción
])
def test_maximo_seguro(nums, esperado):
    """Prueba parametrizada para la función maximo_seguro.
    Se incluyen casos con números positivos, negativos, mixtos y flotantes para verificar su correcto funcionamiento.
    """
    assert maximo_seguro(nums) == esperado # Verificamos que el resultado de la función sea igual al valor esperado para cada caso de prueba
    with pytest.raises(ValueError): # Verificamos que se lance una excepción ValueError cuando la lista esté vacía
        maximo_seguro([])
```

**4-Contador simple de palabras**

```python
# codigo correspondiente a la función contar_palabras del fichero funcs.py

def contar_palabras(texto: str) -> dict:
    conteo = {}
    for palabra in texto.lower().split():
        conteo[palabra] = conteo.get(palabra, 0) + 1
    return conteo
```

```python
# codigo de prueba para la función contar_palabras del fichero test_funcs.py

import pytest
from funcs import contar_palabras

@pytest.mark.parametrize("texto, esperado", [
    ("hola mundo", {"hola": 1, "mundo": 1}),
    ("python es genial python", {"python": 2, "es": 1, "genial": 1}),
    ("", {}), # Caso de texto vacío
    ("una palabra", {"una": 1, "palabra": 1}),
    ("hola hola", {"hola": 2}) # Caso de palabras repetidas
])
def test_contar_palabras(texto, esperado):
    """Prueba parametrizada para la función contar_palabras.
    Se incluyen casos con diferentes textos para verificar el conteo correcto de palabras, incluyendo casos con texto vacío y palabras repetidas.
    """
    assert contar_palabras(texto) == esperado
```

**5-Filtrar aprobados**

```python
# codigo correspondiente a la funcion filtrar_aprobados del fichero funcs.py

def filtrar_aprobados(pares):
    resultado = []
    for nombre, nota in pares:
        if nota >= 5:
            resultado.append(nombre)
    return resultado
```

```python
# codigo correspondiente a la funcion filtrar_aprobados del fichero test_funcs.py

import pytest
from funcs import filtrar_aprobados

@pytest.mark.parametrize("pares, esperado",  [
    # Caso 1: Mezcla y verificación de orden
    ([("Ana", 8), ("Luis", 3), ("Marta", 6)], ["Ana", "Marta"]),
    
    # Caso 2: Todos aprobados
    ([("Carlos", 10), ("Sofía", 5)], ["Carlos", "Sofía"]),
    
    # Caso 3: Todos suspensos (Lista vacía)
    ([("Pedro", 2), ("Lucía", 4.5)], []),
    
    # Caso 4: Límite exacto y tipos (float/int)
    ([("Hugo", 4.9), ("Julia", 5.0), ("Leo", 7)], ["Julia", "Leo"]),
])

def test_filtrar_aprobados(pares, esperado):
    """Prueba parametrizada para verificar la función filtrar_aprobados con diferentes casos de entrada.
    Cada caso incluye una lista de pares (nombre, nota) y la lista esperada de nombres aprobados.
    """
    assert filtrar_aprobados(pares) == esperado #la aserción verifica el orden de los nombres en la lista de resultados, asegurando que se mantenga el orden original de los pares. 
```

**6-Normalizar email**

```python
# codigo correspondiente a la función normalizar_email del fichero funcs.py

def normalizar_email(email: str):
    if email.count('@') != 1:
        raise ValueError('email inválido')
    usuario, dominio = email.split('@')
    return usuario.lower(), dominio.lower()
```

```python
# codigo de prueba para la función normalizar_email del fichero test_funcs.py

import pytest
from funcs import normalizar_email

@pytest.mark.parametrize('email, resultado_esperado', [
    ('juan@ejemplo.com', ('juan', 'ejemplo.com')),
    ('MARIA@EJEMPLO.COM', ('maria', 'ejemplo.com')),
    # Para el error, pasamos la clase de la excepción
    ('usuario@dominio.com@otro.com', ValueError),
])
def test_normalizar_email(email, resultado_esperado):
    """Prueba la función normalizar_email con diferentes casos.
    - Para casos normales, verifica que el resultado sea el esperado.
    - Para casos de error, verifica que se lance la excepción ValueError.
    """
    if resultado_esperado == ValueError: # Si el resultado esperado es una excepción (ValueError)
        with pytest.raises(ValueError):
            normalizar_email(email)
    else:
        # Si es un caso normal
        assert normalizar_email(email) == resultado_esperado
```

**7-Factorial iterativo con control de errores**

```python
# codigo correspondiente a la función factorial del fichero funcs.py

def factorial(n: int) -> int:
    if n < 0:
        raise ValueError('n negativo')
    resultado = 1
    for i in range(2, n+1):
        resultado *= i
    return resultado
```

```python
# codigo de prueba para la función factorial del fichero test_funcs.py

import pytest
from funcs import factorial

@pytest.mark.parametrize('n, esperado', [
    (0, 1), 
    (1, 1), 
    (5, 120)
    ])
def test_factorial_exitos(n, esperado):
    """Prueba que la función factorial devuelve el resultado esperado para casos exitosos."""
    assert factorial(n) == esperado

def test_factorial_error():
    """Prueba que la función factorial lanza un error para valores negativos."""
    with pytest.raises(ValueError):
        factorial(-1)
```

**8-Media aritmética**

```python
# codigo correspondiente a la función media del fichero funcs.py

def media(nums):
    if not nums:
        raise ValueError('lista vacía')
    return sum(nums) / len(nums)
```

```python
# codigo de prueba para la función media del fichero test_funcs.py

import pytest
from funcs import media

# Pruebas parametrizadas para casos de éxito
@pytest.mark.parametrize("lista, esperado", [
    ([1, 2, 3], 2.0),
    ([10, 20], 15.0),
    ([1.5, 2.5, 3.5], 2.5),
    ([1/3, 2/3], 0.5), # Ejemplo para approx
    ([7], 7.0)
])
def test_media_valores(lista, esperado):
    """Verifica que la función media calcule correctamente el promedio de una lista de números
        Usamos pytest.approx para evitar errores de precisión decimal"""
    assert media(lista) == pytest.approx(esperado)

def test_media_lista_vacia():
    """Verifica que la función media lance un ValueError con el mensaje 'lista vacía' cuando se le pasa una lista vacía."""
    with pytest.raises(ValueError, match='lista vacía'):
        media([])
```

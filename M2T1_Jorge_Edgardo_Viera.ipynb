**2-Índices de cadenas**

```python
texto2 = "Python"
print(texto2[0]) #imprime la primera letra de la cadena
print(texto2[-1]) #imprime la ultima letra de la cadena
```

**3-Extraer partes de una dirección (slicing)**

```python
direccion = "Calle Falsa,123, Springfield"
nombre_calle = direccion[0:11] #usando slicing directo imprimimos solo el nombre d ela calle
print(nombre_calle)
numero = direccion[12:15] #usando slicing directo imprimimos el numero de la calle
print(numero)
ciudad = direccion[17:] #usando slicing directo imprimimos el nombre d ela ciudad
print(ciudad)
```

**4-Identificador único con slicing y concatenación**

```python
nombre = "Jorge"
apellido = "viera"
año_nacimiento = "1983"
extraido = nombre[0:3] + apellido[0:3] + año_nacimiento [-2:] #se concatena los tres slicings y se imprime el resultado de las 3 extracciones
print(extraido)
```

**5-Inmutabilidad: modificar una cadena**

```python
palabra = "robot"
palabra[0] = "R" # descomentar para ver el errordescomentar para ver el error, no se permite modificar una cadena
palabra2 = "R"
palabra_modificada = palabra2 + palabra[1:] # slicing y concatenación para obtener "Robot".
print(palabra_modificada)
```

**6-Analizador de palabras**

```python
texto = "Python es POTENTE. Aprender python es útil. PYTHON abre puertas."
objetivo = "Python"

# Normalización
texto_min = texto.lower()
print(texto_min)
obj_min = objetivo.lower()
print(obj_min)

# Otra opcion de impieza de puntuación y normalización sin que los puntos o comas te estorben
texto_limpio = texto.lower().replace(".", "").replace(",", "")

# Conteos
exactas = texto_min.split().count(obj_min) #Total como palabra exacta (split): 3

subcadenas = texto_min.count(obj_min) #Total como subcadena (count): 3

print(f"Exactas: {exactas} | Subcadenas: {subcadenas}")

#Resaltado
import re
texto_remarcado = re.sub(objetivo, objetivo.upper(), texto, flags=re.IGNORECASE)
print(f"Texto: {texto_remarcado}")


```

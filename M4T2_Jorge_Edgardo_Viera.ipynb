- ### PARTE DE `CSV`

***Ejercicio 1. Lectura y almacenamiento de datos***

En la primer consigna logramos trabajar con el fichero `csv` proporcionado `02_CSV_data.csv` extrayendo sus columnas a listas que fueron creadas mediante el codigo. Ejecutamos un bucle para colocar la informacion correspondiente a cada columna y mediante la ejecucion de una `print()` comprobamos que los datos fueron almacenados exitosamente en su lista correspondiente.

```python


import csv # Importamos la biblioteca csv para manejar archivos CSV
from pathlib import Path # Importamos Path para manejar rutas de archivos de manera más fácil

# 1. Creamos las listas vacías para cada columna numérica

mpg = []
cylinders = []
displacement = []
horsepower = []
weight = []
acceleration = []
year = []
origin = []

# 2. Abrimos el archivo usando solo su nombre entre comillas
with open('02_CSV_data.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    
     # Guardamos la primera fila (los nombres de las columnas)
    headers = next(reader) 
    
    # Agregamos el nombre correspondiente como primer elemento de cada lista
    mpg.append(headers[0])
    cylinders.append(headers[1])
    displacement.append(headers[2])
    horsepower.append(headers[3])
    weight.append(headers[4])
    acceleration.append(headers[5])
    year.append(headers[6])
    origin.append(headers[7])
    # Aqui la conversion de valores es por simple prolijidad y claridad de la devolucion de datos
    # en la consola y que se pueda usar posteriormente para graficar o hacer calculos sin necesidad de convertir cada vez
    for row in reader:
        mpg.append(float(row[0])) # ( SPOILER EJERCICIO 2) usamos float para convertir ya que CSV solo almacena texto
        cylinders.append(int(row[1])) # usamos int para convertir ya que CSV solo almacena texto
        displacement.append(float(row[2]))
        horsepower.append(float(row[3]))
        weight.append(float(row[4]))
        acceleration.append(float(row[5]))
        year.append(int(row[6]))
        origin.append(int(row[7]))

# Ejemplo para revisar que funcionó correctamente la lectura de datos y el almacenamiento en las listas
print("FASE DE PRUEBA - Imprimiendo los primeros 5 datos de cada lista:\n")
print("Datos de MPG:", mpg[:5])
print("Datos de Cylinders:", cylinders[:5])
print("Datos de Displacement:", displacement[:5])
print("Datos de Horsepower:", horsepower[:5])
print("Datos de Weight:", weight[:5])
print("Datos de Acceleration:", acceleration[:5])
print("Datos de Year:", year[:5])
print("Datos de Origin:", origin[:5])
```

***Ejercicio 2. Limpieza y conversión de datos***

En este codigo vamos preparando el terreno para quedarnos puramente con los datos, pa ello, eliminamos los encabezados (nombres de columnas) con `pop`
y tambien nos quedamos con los datos en formato correspondiente para ejecutar operaciones y calculos fututos (`int` y `float`) ya que CVS solo procesa todo en `str`. Para obtener esta conversion a valores que podamos usar, nos hacemos del uso de `list` y `map`.

- Por ejemplo: `map(float, mpg)`: Toma la función float y se la aplica de golpe a cada uno de los elementos de la lista `mpg`.
- `list(...)`: Convierte el resultado de `map` (que es un objeto especial) de vuelta en una lista normal de Python para que podamos usarla.
- `type(...)`: Imprimir el tipo de dato (saldrá <class 'float'>) es la mejor forma de comprobar y demostrar que los datos ya no son texto.

```python
# Eliminamos el primer elemento de cada lista para eliminar el encabezado (nombres de columnas)
mpg.pop(0) # Usamos pop(0) para eliminar el primer elemento de la lista, que es el encabezado, y así quedarnos solo con los datos numéricos
cylinders.pop(0)
displacement.pop(0)
horsepower.pop(0)
weight.pop(0)
acceleration.pop(0)
year.pop(0)
origin.pop(0)

# Convertimos los datos de string a float usando map y list 
mpg = list(map(float, mpg))
cylinders = list(map(float, cylinders))
displacement = list(map(float, displacement))
horsepower = list(map(float, horsepower))
weight = list(map(float, weight))
acceleration = list(map(float, acceleration))
year = list(map(float, year))
origin = list(map(float, origin))

# Imprimimos nuevamente los primeros 5 datos de cada lista para verificar que se eliminaron los encabezados y se convirtieron a números
print("\nFASE DE PRUEBA - Imprimiendo los primeros 5 datos de cada lista después de eliminar encabezados y convertir a números:\n")
print("Primeros datos de MPG sin encabezados:", mpg[:5])
print("Tipos de datos en MPG:", type(mpg[0]))
```

***Ejercicio 3. Análisis estadístico***

```python
# Utilizamos la biblioteca statistics para calcular media, mediana y desviación estándar
import statistics as stats

# Agrupamos todas las listas en un diccionario para poder recorrerlas juntas
datos = {
    "MPG": mpg,
    "Cylinders": cylinders,
    "Displacement": displacement,
    "Horsepower": horsepower,
    "Weight": weight,
    "Acceleration": acceleration,
    "Year": year,
    "Origin": origin
}

print("=== ANÁLISIS ESTADÍSTICO ===")

# Recorremos el diccionario y calculamos todo de golpe para cada lista
for nombre, lista in datos.items():
    print(f"\n--- Lista: {nombre} ---")
    print(f"  Observaciones:       {len(lista)}") # Imprime la cantidad de datos en la lista usando len()
    print(f"  Mínimo / Máximo:     {min(lista)} / {max(lista)}") # Imprime el valor mínimo y máximo de la lista usando min() y max()
    print(f"  Media:               {stats.mean(lista):.2f}") # Imprime la media de la lista usando stats.mean()
    print(f"  Mediana:             {stats.median(lista):.2f}") # Imprime la mediana de la lista usando stats.median()
    print(f"  Desviación Estándar: {stats.stdev(lista):.2f}") # Imprime la desviación estándar de la lista usando stats.stdev()
```

***Ejercicio 4. Creación y análisis de un JSON***

Dado el String proporcionado por la consigna se crea detalladamente el diccionario que reproduce la estructura fielmente lo solicitado y se almacena en una variable mostrando tambien, su tipo y contenido haciendo uso de `json.dumps`
Luego cumplimos con el resto de la consigna creando las dos variables solicitadas y realizando calculos con ellas.

```python
# Importamos la libreria de json para guardar los resultados en un archivo JSON
import json


# Creamos un diccionario que reproduzca exactamente la estructura.
datos_proyecto = {
    "jefe_proyecto": {
        "Nombre": "Juan",
        "Edad": 28,
        "Experiencia": ["Gestion", "Finanzas", "Bases de datos"],
        "Residencia": "Madrid",
        "HorasProyecto": 3500
    },
    "empleados": [
        {
            "Nombre": "Elena",
            "Edad": 26,
            "Experiencia": ["JavaScript", "Python"],
            "Residencia": "Madrid",
            "HorasProyecto": 500
        },
        {
            "Nombre": "Luis",
            "Edad": 31,
            "Experiencia": ["Django", "Flask", "Pyramid"],
            "Residencia": "Barcelona",
            "HorasProyecto": 1100
        }
    ]
}

# Almacenamos en una variable y mostramos su TIPO y CONTENIDO
print("\n=== FICHA DEL DICCIONARIO ===")
print("Tipo de la variable:", type(datos_proyecto))
print("\nContenido del diccionario (Formato Python):")
# Usamos un bucle simple para mostrar el contenido de forma ordenada sin usar JSON
for clave, valor in datos_proyecto.items():
    print(f"-> {clave}: {valor}")

# Creamos las dos variables solicitadas
horas_jefe = 0
horas_empleados = 0

# Extraemos las horas del jefe de proyecto
horas_jefe = datos_proyecto["jefe_proyecto"]["HorasProyecto"]

# Extraemos las horas de cada empleado y las sumamos con un bucle
for empleado in datos_proyecto["empleados"]:
    horas_empleados += empleado["HorasProyecto"]

# Mostramos el total de horas del proyecto
total_horas = horas_jefe + horas_empleados

# Imprimimos los resultados
print("\n=== RESULTADOS DE LAS OPERACIONES ===")
print(f"Horas del jefe de proyecto: {horas_jefe}")


```

***Ejercicio 5. Conversión a JSON String***

Dado el diccionario del codigo anterior lo convertimos a formato `JSON String`. Cumpliendo con el requerimiento de mostrar su contenido en dicho formato por consola.
**¿Para qué sirve este formato?**: El formato JSON String (que en Python es una cadena de texto o 'str') tiene como función principal el intercambio y almacenamiento de datos.

*Sus ventajas principales son:*
-  Es universal: Un diccionario de Python solo lo entiende Python. Al transformarlo a texto JSON, cualquier otro lenguaje (JavaScript, PHP o Java) puede leerlo e interpretarlo sin problemas.
- Es ligero: Al ser texto plano, ocupa muy poco espacio, lo que lo hace ideal para enviar información a través de la web (APIs).
- Es legible: Mantiene una estructura limpia y ordenada que puede ser leída y editada fácilmente tanto por computadoras como por humanos.



```python
# Ahora convertimos  el diccionario anterior a formato JSON String.
datos_json = json.dumps(datos_proyecto, indent=4)

# Mostramos su tipo y contenido para verificar que se convirtió correctamente a JSON String
print("\n=== FICHA DEL JSON STRING ===")
print("Tipo de la variable:", type(datos_json))
print("Contenido:")
print(datos_json)
```

***Ejercicio 6. Corrección de errores en un JSON***

 Para empezar, primero necesitamos una estructura base en Python (un diccionario) que represente el grupo de superhéroes con esos errores que menciona la consigna.
 - `.append()`: Lo usamos para añadir un elemento nuevo al final de una lista (como el poder "Volar").
 - Las posiciones [0], [1], [2]: En Python las listas empiezan a contar desde el `0`. Por eso Superman es el 0, Batman el `1` y Wonder Woman el `2`.
 - `ensure_ascii=False`: Este parámetro dentro de `json.dumps()` sirve para que las palabras con tildes (como "Inteligencia" o "Súper") se sigan viendo bien en español y no se transformen en códigos raros.
 

```python
import json

liga_de_la_justicia = {
    "miembros": [
        {
            "nombre": "Superman",
            "edad": 30,
            "poderes": ["superfuerza", "velocidad"] # Falta Volar
        },
        {
            "nombre": "Batman",
            "edad": 350,  # Edad incorrecta
            "poderes": ["inteligencia", "Rayos en los ojos", "Dinero"] # Rayos en los ojos no va
        },
        {
            "nombre": "Wonder Woman",
            "edad": 5000,
            "poderes": ["superfuerza", "Brazaletes"] # Falta Lazo de la verdad
        }
    ]
}

# Correción de errores en los datos

liga_de_la_justicia["miembros"][0]["poderes"].append("Volar") # Agregamos el poder faltante a Superman
liga_de_la_justicia["miembros"][1]["edad"] = 35 # Corregimos la edad de Batman
liga_de_la_justicia["miembros"][1]["poderes"].remove("Rayos en los ojos") # Eliminamos el poder incorrecto de Batman
liga_de_la_justicia["miembros"][2]["poderes"].append("Lazo de la verdad") # Agregamos el poder faltante

# Convertimos el diccionario corregido a JSON String para mostrarlo
liga_json = json.dumps(liga_de_la_justicia, indent=4, ensure_ascii=False)
print("\n=== LIGA DE LA JUSTICIA CORREGIDA ===")
print(liga_json)
```

***Ejercicio 7. Reorganización del JSON***

**La reorganizacion consistió en los siguientes pasos:**

- *Aislamiento de datos*: Se extrajeron los objetos (diccionarios) de cada superhéroe en variables independientes `(superman, batman, wonder_woman) `haciendo uso de sus índices o posiciones indexadas.

- *Cambio de estructura*: Se transformó el valor de la clave `"miembros"`, que dejó de ser una lista para convertirse en un diccionario contenedor.

- *Categorización*: Dentro de este nuevo diccionario, se crearon dos nuevas claves de tipo lista `("miembrosActivos" y "miembrosInactivos")`. Esto permite clasificar y agrupar a los personajes según su estado actual.

```python
# Ahora extraemos los datos de cada miembro en variables separadas para poder usarlos posteriormente
superman = liga_de_la_justicia["miembros"][0] # Obtenemos el diccionario de Superman
batman = liga_de_la_justicia["miembros"][1] # Obtenemos el diccionario de Batman
wonder_woman = liga_de_la_justicia["miembros"][2] # Obtenemos el diccionario de Wonder Woman

# Modificamos la estructura para que miembros tenga miembrosActivos y miembrosInactivos
liga_de_la_justicia ["miembros"] = {
    "miembrosActivos": [superman, wonder_woman], 
    "miembrosInactivos": [batman] 
}

# Convertimos la nueva estructura a JSON String para mostrarla
liga_json_modificada = json.dumps(liga_de_la_justicia, indent=4, ensure_ascii=False)
print("\n=== LIGA DE LA JUSTICIA MODIFICADA ===")
print(liga_json_modificada)
```

***Ejercicio 8. Consumo de una API***

*Accedemos a una API proporcionada:*

- Mostramos el tipo con `type(data)`.
- Mostramos los datos completos bien formateados con `json.dumps()`.
- Mostramos la cantidad total usando la clave `data['number']`.
- Recorremos la lista con un bucle for para imprimir el nombre y la nave de cada astronauta.

```python
# NOTA:para que este codigo funcione anteriormente en el terminal se debe instalar la biblioteca requests usando el comando: pip install requests

import requests
import json

# 1. Nos conectamos a la API y descargamos los datos en tiempo real
response = requests.get("http://api.open-notify.org/astros.json")
data = response.json()

# mostramos el tipo de dato que recibimos 
print("\n=== FICHA DE LOS DATOS RECIBIDOS DE LA API ===")
print("Tipo de dato recibido:", type(data))
print("\nContenido de los datos recibidos:")
print(json.dumps(data, indent=4, ensure_ascii=False))

# recorremos el diccionario para mostrar número de personas en el espacio y el nombre y nave de cada persona
print("\n=== PERSONAS EN EL ESPACIO EN ESTE MOMENTO ===")
print(f"Número total de personas en el espacio: {data['number']}")
for person in data['people']:
    print(f"Nombre: {person['name']}, Nave: {person['craft']}")
```

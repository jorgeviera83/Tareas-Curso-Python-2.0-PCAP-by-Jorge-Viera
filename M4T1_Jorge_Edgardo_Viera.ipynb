- ### Ejercicio 1. Información de un fichero

*Los sucesos más importantes ocurridos en este código informativo sobre un fichero es que al usar `with`, Python se encarga de cerrar el fichero automáticamente cuando termina el bloque de código, incluso si ocurre un error inesperado por el camino. También, dentro del bloque `with` nos mostrará `False` (porque estamos trabajando en él), y justo al salir del bloque pasará a ser `True`.*

**Esto hace que nuestro código sea más seguro y limpio**

```python
from pathlib import Path

# Definimos la ruta del fichero usando Path
ruta_fichero = Path("pruebas.txt")

# Comprobamos si el archivo existe
if not ruta_fichero.exists():
    print(f"Error: El archivo no existe en la ruta actual: {ruta_fichero.absolute()}")
else:
    # Abrimos el fichero usando "with" (gestión automática de cierre)
    with ruta_fichero.open("r", encoding="utf-8") as fichero:
        # Mostramos la información requerida MIENTRAS está abierto
        print(f"Nombre del fichero: {fichero.name}")
        print(f"¿El fichero está cerrado dentro del bloque?: {fichero.closed}")
        print(f"Modo de apertura: {fichero.mode}")

        # Listamos todo el contenido
        print("\n--- Lista con el contenido del fichero ---")
        print(fichero.read())
    
    # Comprobamos que "with" lo cerró automáticamente al salir del bloque
    print("--------ANEXO DE SEGURIDAD---------")
    print(f"¿El fichero está cerrado fuera del bloque?: {fichero.closed}")


    
```

- ### Ejercicio 2. Lectura del fichero

*Aquí en este código con el afán de cumplir las consignas del ejercicio y de elaborar un código legible, simple y lo más profesional posible, aparte de utilizar `readline` nos hacemos del metodo `seek`.*

- *Este método reinicia el puntero de lectura y lo vuelve a colocar en la posición cero (el inicio del archivo).*
- *Evita tener que cerrar y reabrir el archivo con otro with open, lo cual consume menos recursos del ordenador y hace que tu script sea más eficiente.*

```python

# leemos la primera línea del fichero
with ruta_fichero.open("r", encoding="utf-8") as fichero:
    print("--- Lectura de una sola línea ---")
    primera_linea = fichero.readline()
    print(primera_linea.strip())

    # retrocedemos el cursor al inicio del fichero para leerlo de nuevo
    fichero.seek(0)
    # el bucle for lee el fichero línea a línea hasta el final
    print("\n--- Lectura del fichero línea a línea ---")
    for linea in fichero:
        print(linea.strip())
```

- ### Ejercicio 3. Escritura concatenando contenido

*Cumpliendo con las consignas se utiliza append `"a"` y `with` para lograr las siguientes acciones,*
- *El primer `with` abre, escribe y cierra el fichero al salir del bloque.*
- *El segundo `with` lo vuelve a abrir en modo lectura para mostrar en la pantalla todo el texto, confirmando visualmente que la nueva línea ya forma parte de nuestro archivo `pruebas.txt`.



```python
# Abrimos el fichero en modo "append" (añadir al final)
with ruta_fichero.open("a", encoding="utf-8") as fichero:
    # Escribimos la nueva línea (añadimos \n al final para futuros textos)
    fichero.write("Esta es una nueva línea añadida al final del fichero.\n")

# El bloque "with" se encarga de CERRAR el fichero de forma segura aquí.

# Volvemos a abrir el fichero en modo lectura ("r") para COMPROBAR
print("--- Comprobación del contenido actualizado ---")
with ruta_fichero.open("r", encoding="utf-8") as fichero:
    for linea in fichero:
        print(linea.strip())
```

- ### Ejercicio 4. Escritura sobrescribiendo contenido

*A diferencia del codigo del ejercicoo anterior aqui utilizamos `"w"` que crea o sobrescribe el fichero. Tambien,*
- *El primer `with` abre, escribe y cierra el fichero al salir del bloque.*
- *El segundo `with` lo vuelve a abrir en modo lectura para mostrar en la pantalla todo el texto, confirmando visualmente que **el nuevo contenido** ya forma parte de nuestro archivo `pruebas.txt`.

```python
# Abrimos el fichero en modo escritura sobrescribiendo su contenido.
with ruta_fichero.open("w", encoding="utf-8") as fichero:
    fichero.write("Este es el nuevo contenido del fichero.\nSobrescribiendo el anterior.")

# Comprobamos que el contenido se ha sobrescrito correctamente
print("--- Comprobación del contenido sobrescrito ---")
with ruta_fichero.open("r", encoding="utf-8") as fichero:
    for linea in fichero:
        print(linea.strip())
```

- ### Ejercicio 5. Procesamiento de datos desde un fichero

*Trabajamos con un nuevo fichero llamado `personas.txt`. Al igual que el primer ejercicio definimos la ruta y nos aseguramos de su existencia.*
*Utilizando metodos como `.split()` logramos que nos divida una sola cadena de texto en varias partes independientes y nos devuelve una lista con esos elementos. Al igual que `.strip()` en cada campo: Limpia cualquier espacio sobrante que pueda quedar alrededor de los nombres o apellidos tras hacer la división.*
* *En la primera fase, lectura, se abre el archivo con with open, extraemos los datos rápidamente, los guardamos en la lista personas y dejamos que Python cierre el archivo.*
* *En la segunda fase, presentación, con el archivo ya cerrado y los datos seguros en la memoria, recorremos la lista para mostrarlos de forma amigable.*

```python


from pathlib import Path

ruta_fichero = Path("personas.txt")

# Lista donde guardaremos los diccionarios de cada persona
personas = []

if not ruta_fichero.exists():
    print(f"Error: El archivo no existe en la ruta actual: {ruta_fichero.absolute()}") # Comprobamos que el archivo existe antes de intentar leerlo 
else:
    with ruta_fichero.open("r", encoding="utf-8") as fichero: # Abrimos el fichero en modo lectura y con codificación UTF-8
        for linea in fichero: # Iteramos por cada línea del fichero
            # Limpiamos espacios y saltos de línea
            linea_limpia = linea.strip()
            # Separamos los datos por el punto y coma
            datos = linea_limpia.split(";")
            
            # Creamos el diccionario asignando cada campo a su posición
            persona = {
                "id": datos[0],
                "nombre": datos[1],
                "apellido": datos[2],
                "nacimiento": datos[3]
            }
            
            # Añadimos el diccionario a la lista
            personas.append(persona)

    print("--- Información de personas ---")
    for persona in personas:
        print(f"(id={persona['id']}) {persona['nombre']} {persona['apellido']} => {persona['nacimiento']}")
```

**1-Orden de los números**

```python
num_1 = 1 #creamos la variable num_1 y le asignamos el valor 1 para no usar una lista

if num_1 == 0:
    print("Error: El primer número no puede ser 0.") #aqui nos dara error si usaramos el numero 0
else:
    num_2 = 2 #creamos la variable num_2 y le asignamos el valor 2
    num_3 = 3 #creamos la variable num_3 y le asignamos el valor 3

    if num_1 < num_2 < num_3: #aqui comparamos los numeros para ver si estan en orden ascendente
        print("Los números están en orden ascendente.")
    else:
        print("Los números NO están en orden ascendente.") #alternativamente, si no estan en orden ascendente, nos dara este mensaje
```

**2-Orden numérico con listas**

```python
numeros = [1, 2, 3] #creamos una lista llamada numeros con los valores 1, 2 y 3
#aqui nos dara error si usamos el numero 0 en la lista, ya que no se puede comparar con los otros numeros
if numeros[0] == 0:
    print("Error: El primer número no puede ser 0.")
else:
    if numeros[0] < numeros[1] < numeros[2]: #aqui comparamos los numeros de la lista para ver si estan en orden ascendente
        print("Los números están en orden ascendente.")
    else:
        print("Los números NO están en orden ascendente.") #alternativamente, si no estan en orden ascendente, nos dara este mensaje
```

**3-Sumatorio de los primeros N números**

```python
n = int(input("Introduce el número límite (N): ")) # aqui pedimos al usuario que introduzca un numero limite (5) para el sumatorio, y lo convertimos a entero
suma = 0 #creamos la variable suma y le asignamos el valor 0 para poder ir sumando los numeros desde 0 hasta n-1

for i in range(n): #creamos un bucle for que va desde 0 hasta n-1, ya que n es 5, el bucle se ejecutara con i tomando los valores 0, 1, 2, 3 y 4
    suma += i  # aqui sumamos el valor de i a la variable suma en cada iteracion del bucle
print(f"El sumatorio es: {suma}") #aqui imprimimos el resultado del sumatorio, que sera la suma de los numeros desde 0 hasta n-1
```

**4-Contador de letras "a"**

```python
contador = 0 # asignamos a la variable contador el valor 0 para contar las veces que aparece la letra 'a'

while True:
    letra = input("Introduce letras una por una. Para terminar, escribe un punto (.).: ").lower() # pedimos al usuario que introduzca una letra y la convertimos a minúscula para contar tanto 'A' como 'a'
    
    if letra == '.':# si el usuario introduce un punto, salimos del bucle con break
        break
    if letra == 'a': # si la letra introducida es 'a', incrementamos el contador en 1
        contador += 1

print(f"La letra 'a' apareció {contador} veces.") # finalmente, imprimimos el resultado del contador, que es el número de veces que apareció la letra 'a' en las entradas del usuario
```

**5-Contador de letras "a" en palabras**

```python
total_letra_a = 0

while True:
    palabra = input("Introduce una palabra (o '.' para terminar): ")
    
    if palabra == ".": # si el usuario introduce un punto, salimos del bucle con break
        break
    
    # Contamos 'a' minúsculas y 'A' mayúsculas
    total_letra_a += palabra.lower().count('a')

print(f"La letra 'a' apareció {total_letra_a} veces.") # finalmente, imprimimos el resultado del contador, que es el número de veces que apareció la letra 'a' en las palabras introducidas por el usuario
```

**6-Sumatorio de números pares decrecientes**

```python
# Calculamos el sumatorio de los números pares entre 1000 y 73 en orden descendente  
suma_pares = 0 #creamos la variable suma_pares y le asignamos el valor 0 para ir sumando los numeros pares encontrados en el bucle for
for n in range(1000, 72, -1): # creamos un bucle for que va desde 1000 hasta 73 en orden descendente, con un paso de -1
    if n % 2 == 0: # si el número n es par (es decir, el resto de n dividido por 2 es 0), entonces lo sumamos a la variable suma_pares
        suma_pares += n
print(f"El sumatorio de los números pares entre 1000 y 73 en orden descendente es: {suma_pares}")
```

**7-Lista de números: análisis y operaciones**

```python
#solicitar la cantidad de números a ingresar
cantidad = int(input("¿Cuántos números deseas ingresar? "))

#leer los números y almacenarlos en una lista
numeros = []
for _ in range(cantidad):
    numero = int(input("Introduce un número: "))
    numeros.append(numero)
    print(f"Número {numero} agregado a la lista.")
#ordena la lista de números en orden ascendente
numeros.sort()
print(f"Los números ordenados de menor a mayor son: {numeros}")
#calcular el promedio de los números
promedio = sum(numeros) / len(numeros)
print(f"El promedio de los números ingresados es: {promedio}")
#identificar el mayor y el menor número y la posición de cada uno en la lista
mayor = max(numeros) #calculamos el mayor número de la lista usando la función max()
menor = min(numeros) #calculamos el menor número de la lista usando la función min()
posicion_mayor = numeros.index(mayor) #obtenemos la posición del mayor número en la lista usando el método index()
posicion_menor = numeros.index(menor) #obtenemos la posición del menor número en la lista usando el método index()
print(f"El mayor número es: {mayor} y está en la posición: {posicion_mayor}")
print(f"El menor número es: {menor} y está en la posición: {posicion_menor}")
```

**8-Menú compra-venta**

```python
compras = []
ventas = []
alquileres = []

# El menú se imprime solo una vez al inicio
print("--- MENU COMPRA-VENTA ---")
print("1. Comprar")
print("2. Vender")
print("3. Alquilar")
print("4. Mostrar listas de datos")
print("5. Salir")

while True:
    opcion = input("\nSeleccione una opción (1-5): ") #pedimos al usuario que seleccione una opción del menú

    if opcion == "1":
        item = input("¿Qué desea comprar?: ") #si el usuario selecciona la opción 1, le preguntamos qué desea comprar y lo almacenamos en la variable item
        compras.append(item)
    
    elif opcion == "2":
        item = input("¿Qué desea vender?: ") #si el usuario selecciona la opción 2, le preguntamos qué desea vender y lo almacenamos en la variable item
        ventas.append(item)
        
    elif opcion == "3":
        item = input("¿Qué desea alquilar?: ") #si el usuario selecciona la opción 3, le preguntamos qué desea alquilar y lo almacenamos en la variable item
        alquileres.append(item)
        
    elif opcion == "4": #si el usuario selecciona la opción 4, mostramos las listas de compras, ventas y alquileres
        print(f"\nCompras: {compras}")
        print(f"Ventas: {ventas}")
        print(f"Alquileres: {alquileres}")
        
    elif opcion == "5": #si el usuario selecciona la opción 5, finalizamos el programa con un mensaje de despedida y salimos del bucle con break
        print("Finalizando programa...")
        break
        
    else:
        print("Opción no válida.")

```

**9-Programa de login con reintentos**

```python
while True:
    # 1. Solicitar y guardar la contraseña inicial
    password_guardada = input("\n[CONFIGURACIÓN] Establece tu nueva contraseña: ")
    intentos = 0 #se establece un minimo de intentos
    max_intentos = 5 #se establece un maximo de intentos para el login
    acceso_concedido = False #variable para controlar si el acceso fue concedido o no

    while intentos < max_intentos:
        # 2. Pedir contraseña para iniciar sesión
        login_input = input("Introduce tu contraseña para iniciar sesión: ")
        
        # 3. Validar contraseña
        if login_input == password_guardada:
            print("¡Login correcto!")
            acceso_concedido = True
            break
        else:
            # 4. Contador de intentos fallidos
            intentos += 1
            print(f"Login fallido. Intento {intentos}")

    # 5. Si llega a 5 intentos, el bucle externo reinicia el proceso
    if not acceso_concedido:
        print("\nHas agotado los 5 intentos. Reiniciando configuración...")
    else:
        # Si el login fue correcto, terminamos el programa completamente
        break
```

**10-Verificar aprobados y suspendidos con un diccionario**

```python
notas = {"Ana": 8, "Luis": 4, "Marta": 6, "Pedro": 3}
# aprobado si la nota es mayor o igual a 5
for alumno, nota in notas.items(): #recorremos el diccionario de notas usando el método items() para obtener tanto el nombre del alumno como su nota
    if nota >= 5: #si la nota es mayor o igual a 5, el alumno esta aprobado
        print(f"{alumno} está aprobado con una nota de {nota}.")
    else: #si la nota es menor a 5, el alumno esta suspendido
        print(f"{alumno} está suspendido con una nota de {nota}.")
```

### ***Apartado 1. Trabajando con NumPy***

**NOTA**: Aunque no es solicitado en la consiga, al estar en fase de prueba y desarrollo, utilizo print para mostrar resultados en la consola.

En este codigo ademas de usar las funciones de *NumPy*, usamos cuatro opciones de comparación que son justo las más representativas:

- (==) para ver la comparación detallada elemento a elemento
- (array_equal) como la forma estándar y más directa de saber si dos arrays son idénticos
- (.all()) como alternativa manual al mismo resultado
- (allclose) que es la opción más robusta cuando trabajamos con decimales.

```python
import numpy as np

# Creamos un array con valores del 10 al 100.
array = np.arange(10, 101)

# Ahora invertimos el array.
array_invertido = array[::-1]

# Imprimimos el array original y el invertido.
print("Array original:")
print(array)
print("Array invertido:")
print(array_invertido)

# Creamos un array de dimensión 3x3 con valores del 0 al 8.
array_3x3 = np.arange(9).reshape(3, 3)
print("Array 3x3:")
print(array_3x3)

# Ahora creamos un array identidad de 6x6.
array_identidad = np.eye(6)
print("Array identidad 6x6:")
print(array_identidad)

# Creamos un array con valores aleatorios de forma 3x3x3.
array_aleatorio = np.random.rand(3, 3, 3)
print("Array aleatorio 3x3x3:")
print(array_aleatorio)

# Creamos dos arrays aleatorios A y B de la misma forma.
A = np.random.randint(0, 10, (3, 3)) #randint genera números enteros aleatorios en un rango especificado
B = np.random.randint(0, 10, (3, 3))

print("Array A:")
print(A)
print("Array B:")
print(B)

# Comprobamos si son iguales con distintas opciones:

# 1. Comparación elemento a elemento (devuelve un array de booleanos)
print("Comparación elemento a elemento:")
print(A == B)

# 2. np.array_equal -> compara forma y valores, devuelve un único booleano
print("¿Son iguales (array_equal)?:", np.array_equal(A, B))

# 3. all() sobre la comparación elemento a elemento
print("¿Son iguales (all)?:", (A == B).all())

# 4. np.allclose -> compara con tolerancia, útil para arrays de decimales
print("¿Son iguales (allclose)?:", np.allclose(A, B))
```

### ***Apartado 2. NumPy: conversión de valores***

**NOTA:** *En el apartado del ejercicio me pide que cree el array **dólares** (con tilde) pero usamos **dolares** sin tilde, lo cual en Python es perfectamente válido y de hecho es buena práctica evitar tildes en nombres de variables*

```python
# Creamos un array de 100 valores aleatorios enteros entre 1 y 1000.
dolares = np.random.randint(1, 1001, 100)
print("Array aleatorio de 100 valores entre 1 y 1000:")
print(dolares)

# Creamos otro array llamado Euro con los mismos valores pero convertidos a euros (1 USD = 0,874413 EUR) redondeado a 2 decimales.
euros = np.round(dolares * 0.874413, 2)
print("Array convertido a euros:")
print(euros)

# Aunque prefiero mantener un codigo mas simple y limpio como el de arriba
# tambien muetro las otras dos conversiones segun se puede leer en el enunciado del ejercicio

euros_v1 = np.round(dolares * 0.874413, 2)
euros_v2 = np.round(dolares / 1.14362, 2)

print(np.allclose(euros_v1, euros_v2)) # da como resultado False, ya que no son exactamente iguales, aunque si muy parecidos.
```

### ***Apartado 3. pandas y dataset tips de seaborn***

```python
import pandas as pd
import seaborn as sns

# Cargamos el dataset de ejemplo "tips" de Seaborn.
df = sns.load_dataset("tips")

# Mostramos los 5 primeros registros.
print("Primeros 5 registros del dataset 'tips':\n")
print(df.head())

# Mostraamos el numero total de registros y columnas.
print("\nNúmero total de registros y columnas:")
print("Número total de registros:", len(df)) 

# Calculamos la propina media y la cuenta media.
print("\nPropina media:", df["tip"].mean()) #mean() calcula la media de los valores de la columna especificada
print("Cuenta media:", df["total_bill"].mean())

# Creamos la columna pct_tip y mostramos las 5 filas con mayor porcentaje.
df["pct_tip"] = df["tip"] / df["total_bill"] * 100
print("\n5 Filas con mayor porcentaje de propina:")
print(df.nlargest(5, "pct_tip"))

# Mostramos las 10 cuentas más altas con su propina y porcentaje.
print("\n10 Cuentas más altas con su propina y porcentaje:")
print(df.nlargest(10, "total_bill")[["total_bill", "tip", "pct_tip"]])

# Mostramos el promedio de propina por día ordenado de mayor a menor.
print("\nPromedio de propina por día (ordenado de mayor a menor):")
print(df.groupby("day")["tip"].mean().sort_values(ascending=False)) #groupby() agrupa los datos por la columna especificada, 
# mean() calcula la media de cada grupo y sort_values() ordena los valores

# Mostramos el promedio de pct_tip por sexo y por condición de fumador.
print("\nPromedio de pct_tip por sexo y por condición de fumador:")
print(df.groupby(["sex", "smoker"])["pct_tip"].mean()) #groupby() agrupa los datos por las columnas especificadas.

# Indicamos el día y momento en el que se registró la cuenta más alta.
max_bill = df["total_bill"].idxmax() # idxmax() devuelve el índice del valor máximo de la columna especificada
print("\nDía y momento de la cuenta más alta:")
print(df.loc[max_bill, ["day", "time"]]) # loc[] permite acceder a una fila o columna específica del DataFrame por su índice o nombre de columna.

# Mostramos cuántos registros corresponden a Lunch y Dinner.
print("\nNúmero de registros por tipo de comida:")
print(df["time"].value_counts()) # value_counts() cuenta el número de ocurrencias de cada valor en la columna especificada.

# Mostramos la propina media por tamaño de mesa e indicamos cuál deja mayor porcentaje.
print("\nPropina media por tamaño de mesa:")
print(df.groupby("size")["tip"].mean()) # groupby() agrupa los datos por la columna especificada y mean() calcula la media de cada grupo.
print("\nTamaño de mesa que deja mayor porcentaje de propina:")
print(df.groupby("size")["pct_tip"].mean().idxmax()) # idxmax() devuelve el índice del valor máximo de la columna especificada.

# Filtramos mesas de 5 o más personas y calcula su pct_tip medio.
print("\nPromedio de pct_tip para mesas de 5 o más personas:")
print(df[df["size"] >= 5]["pct_tip"].mean()) # filtra las filas del DataFrame donde el tamaño de la mesa es mayor o igual a 5 y calcula la media de la columna pct_tip.

# Calculamos, por día, la facturación total y el total de propinas.
print("\nFacturación total y total de propinas por día:")
facturacion_propinas = df.groupby("day")[["total_bill", "tip"]].sum() # groupby() agrupa los datos por la columna especificada y sum() calcula la suma de cada grupo.
print(facturacion_propinas)

# Calculamos la correlación entre total_bill y tip y comentamos el resultado.
correlacion = df["total_bill"].corr(df["tip"]) # corr() calcula la correlación entre dos columnas.
print("\nCorrelación entre total_bill y tip:")
print(correlacion)
# La correlación es positiva y cercana a 0.67, lo que indica que a medida que aumenta la cuenta total, también tiende a aumentar la propina.

# Exportamos a CSV las columnas indicadas en un archivo tips_resumen.csv.
df[["total_bill", "tip", "pct_tip", "sex", "smoker", "day", "time", "size"]].to_csv("tips_resumen.csv", index=False)
```

### ***Apartado 4. pandas y datasets de IMDb***

*Para obtener un acabado mas profesional y legible al ojo humano, usamos `display()` de IPython para todos los resultados que sean DataFrames o Series (tablas), y dejamos `print()` solo para los textos descriptivos y los valores únicos (números, contadores).*

```python
from IPython.display import display #display() permite mostrar DataFrames de manera más legible en Jupyter Notebook.

# Cargamos los ficheros imdb_titulos.csv y imdb_elenco.csv como DataFrames.
imdb_titulos = pd.read_csv('imdb_titulos.csv', delimiter=',') # delimiter=',' indica que el separador de columnas es la coma, ya que los archivos CSV pueden tener distintos delimitadores.
imdb_elenco = pd.read_csv('imdb_elenco.csv', delimiter=',')

# Mostramos los 5 primeros registros de cada DataFrame.
print("Primeros 5 registros de imdb_titulos:")
display(imdb_titulos.head()) # head() devuelve las primeras 5 filas del DataFrame

print("\nPrimeros 5 registros de imdb_elenco:")
display(imdb_elenco.head())

# Mostramos el número total de registros y columnas de cada DataFrame.
print("\nNúmero total de registros y columnas de imdb_titulos:", imdb_titulos.shape) # shape devuelve una tupla con el número de filas y columnas del DataFrame
print("Número total de registros y columnas de imdb_elenco:", imdb_elenco.shape)

# Mostramos las 5 películas más antiguas.
print("\n5 películas más antiguas:")
display(imdb_titulos.nsmallest(5, "year")[["title", "year"]]) # nsmallest() devuelve las n filas con los valores más pequeños de la columna especificada

# Mostramos las películas cuyo título contenga la palabra "Dracula" y el total de coincidencias.
print("\nPelículas cuyo título contiene 'Dracula':")
dracula_movies = imdb_titulos[imdb_titulos["title"].str.contains("Dracula", case=False, na=False)]
display(dracula_movies[["title", "year"]])
print("Total de coincidencias:", len(dracula_movies))

# Mostramos los 10 títulos más repetidos.
print("\n10 títulos más repetidos:")
display(imdb_titulos["title"].value_counts().head(10)) # value_counts() cuenta el número de ocurrencias de cada valor en la columna especificada y head(10) devuelve las primeras 10 filas del resultado.

# Mostramos la primera película titulada "Romeo and Juliet".
print("\nPrimera película titulada 'Romeo and Juliet':")
romeo_juliet = imdb_titulos[imdb_titulos["title"] == "Romeo and Juliet"].sort_values("year").head(1)
display(romeo_juliet[["title", "year"]])

# Listamos las películas que contengan "Exorcist" ordenadas por año.
print("\nPelículas que contienen 'Exorcist' ordenadas por año:")
exorcist_movies = imdb_titulos[imdb_titulos["title"].str.contains("Exorcist", case=False, na=False)].sort_values("year")
display(exorcist_movies[["title", "year"]])

# Mostramos cuántas películas se hicieron en 1950.
movies_1950 = imdb_titulos[imdb_titulos["year"] == 1950]
print("\nNúmero de películas hechas en 1950:", len(movies_1950))

# Mostramos cuántas películas se hicieron entre 1950 y 1959.
movies_1950_1959 = imdb_titulos[(imdb_titulos["year"] >= 1950) & (imdb_titulos["year"] <= 1959)]
print("Número de películas hechas entre 1950 y 1959:", len(movies_1950_1959))

# Mostramos todos los papeles de la película "The Godfather" y el número total.
print("\nPapeles de la película 'The Godfather':")
godfather_roles = imdb_elenco[imdb_elenco["title"] == "The Godfather"]
display(godfather_roles[["name", "character"]])
print("Número total de papeles en 'The Godfather':", len(godfather_roles))

# Mostramos el elenco ordenado por clasificación n de "Dracula" (1958).
print("\nElenco de 'Dracula' (1958) ordenado por clasificación n:")
dracula_1958_cast = imdb_elenco[(imdb_elenco["title"] == "Dracula") & (imdb_elenco["year"] == 1958)].sort_values("n")
display(dracula_1958_cast[["name", "character", "n"]])

# Mostramos cuántos papeles existen de "Bruce Wayne".
bruce_wayne_roles = imdb_elenco[imdb_elenco["character"].str.contains("Bruce Wayne", case=False, na=False)]
print("\nNúmero de papeles de 'Bruce Wayne':", len(bruce_wayne_roles))

# Mostramos cuántos papeles ha interpretado "Robert De Niro".
robert_de_niro_roles = imdb_elenco[imdb_elenco["name"].str.contains("Robert De Niro", case=False, na=False)]
print("Número de papeles interpretados por 'Robert De Niro':", len(robert_de_niro_roles))

# Listamos los papeles protagonistas (n=1) de Charlton Heston en los años 60.
print("\nPapeles protagonistas (n=1) de Charlton Heston en los años 60:")
charlton_heston_roles_60s = imdb_elenco[
    (imdb_elenco["name"].str.contains("Charlton Heston", case=False, na=False)) &
    (imdb_elenco["n"] == 1) &
    (imdb_elenco["year"] >= 1960) &
    (imdb_elenco["year"] < 1970)
]
display(charlton_heston_roles_60s[["title", "year", "character"]])

# Confirmamos los tipos de valores únicos en la columna "type" del DataFrame imdb_elenco.
print("\nValores únicos en la columna 'type':", imdb_elenco["type"].unique())

# Mostramos cuántos papeles de actores hubo en los años 50.
actores_roles_50s = imdb_elenco[
    (imdb_elenco["n"] == 1) &
    (imdb_elenco["year"] >= 1950) &
    (imdb_elenco["year"] < 1960) &
    (imdb_elenco["type"] == "actor")
]
print("\nNúmero de papeles de actores en los años 50:", len(actores_roles_50s))

# Mostramos cuántos papeles de actrices hubo en los años 50.
actrices_roles_50s = imdb_elenco[
    (imdb_elenco["n"] == 1) &
    (imdb_elenco["year"] >= 1950) &
    (imdb_elenco["year"] < 1960) &
    (imdb_elenco["type"] == "actress")
]
print("Número de papeles de actrices en los años 50:", len(actrices_roles_50s))
```

**1-Función para calcular el área de un rectángulo**

```python
def area_rectangulo(): #función para calcular el área de un rectángulo
    area = base * altura
    return area

#se solicita al usuario que ingrese la base y la altura del rectángulo
base = float(input("Ingrese la base del rectángulo: "))
altura = float(input("Ingrese la altura del rectángulo: "))
#se llama a la función para calcular el área y se muestra el resultado
resultado = area_rectangulo()
print(f"El área del rectángulo es: {resultado}")



```

**3-Función para encontrar el mayor de tres números**

```python
def lee_numero(): #función para leer un número entero del usuario
    numero = int(input("Ingrese un número entero: "))
    return numero
def mayor(num1, num2, num3): #función para encontrar el mayor de tres números
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3
#se solicita al usuario que ingrese tres números enteros
num1 = lee_numero()
num2 = lee_numero()
num3 = lee_numero() 
#se llama a la función para encontrar el mayor y se muestra el resultado
resultado = mayor(num1, num2, num3)
print(f"El mayor de los tres números es: {resultado}")
```

**4-Cálculo del índice de masa corporal (IMC)**

```python
def imc(peso, altura):
    indice = peso / (altura ** 2)
    
    if indice < 18.50:
        clasificacion = "Bajo peso"
    elif 18.50 <= indice < 25.00:
        clasificacion = "Normal"
    elif 25.00 <= indice < 30.00:
        clasificacion = "Sobrepeso"
    else:
        clasificacion = "Obesidad"
    
    return indice, clasificacion

# Entrada de datos
peso = float(input("Ingrese su peso en kilogramos: "))
altura = float(input("Ingrese su altura en metros (con dos decimales): "))

# Desempaquetamos los dos valores que devuelve la función
valor_imc, categoria = imc(peso, altura)

# Mostramos el resultado correctamente
print(f"Su IMC es: {valor_imc:.2f}")
print(f"Clasificación: {categoria}")

```

**5-Login estructurado en funciones**

```python
# Programa de configuración de contraseña con límite de intentos realizado en el ejercicio 9 del Tema 2


while True:
    # 1. Solicitar y guardar la contraseña inicial
    password_guardada = input("\n[CONFIGURACIÓN] Establece tu nueva contraseña: ")
    intentos = 0 #se establece un minimo de intentos
    max_intentos = 5 #se establece un maximo de intentos para el login
    acceso_concedido = False #variable para controlar si el acceso fue concedido o no

    while intentos < max_intentos:
        # 2. Pedir contraseña para iniciar sesión
        login_input = input("Introduce tu contraseña para iniciar sesión: ")
        
        # 3. Validar contraseña
        if login_input == password_guardada:
            print("¡Login correcto!")
            acceso_concedido = True
            break
        else:
            # 4. Contador de intentos fallidos
            intentos += 1
            print(f"Login fallido. Intento {intentos}")

    # 5. Si llega a 5 intentos, el bucle externo reinicia el proceso
    if not acceso_concedido:
        print("\nHas agotado los 5 intentos. Reiniciando configuración...")
    else:
        # Si el login fue correcto, terminamos el programa completamente
        break
```

```python
# Restructuracion del código dividiendolo en funciones

def establecer_password(): #función para establecer la contraseña inicial
    return input("\n[CONFIGURACIÓN] Establece tu nueva contraseña: ")

def pedir_login():#función para pedir la contraseña al usuario para iniciar sesión
    return input("Introduce tu contraseña para iniciar sesión: ")

def validar_password(input_usuario, password_guardada): #función para validar si la contraseña ingresada por el usuario coincide con la contraseña guardada
    return input_usuario == password_guardada

def gestionar_intentos(password_guardada, max_intentos=5): #función para gestionar los intentos de login y reiniciar la configuración si se agotan los intentos
    intentos = 0
    while intentos < max_intentos: #bucle para permitir al usuario intentar ingresar la contraseña hasta un máximo de intentos
        login_input = pedir_login()
        
        if validar_password(login_input, password_guardada):#si la contraseña es correcta, se concede el acceso y se termina el programa
            print("¡Login correcto!")
            return True
        
        intentos += 1
        print(f"Login fallido. Intento {intentos}")
        
    print(f"\nHas agotado los {max_intentos} intentos. Reiniciando configuración...") #si se agotan los intentos, se muestra un mensaje y se reinicia la configuración
    return False

def ejecutar_programa(): #función para ejecutar el programa principal
    while True:#bucle principal que permite reiniciar la configuración si se agotan los intentos
        password_maestra = establecer_password()
        if gestionar_intentos(password_maestra):
            break

# Iniciar el programa
if __name__ == "__main__":
    ejecutar_programa()
```

**6-Función intermedio**

```python
def intermedio(): #función para calcular el número intermedio entre dos números
#se solicita al usuario que ingrese dos números
    num1 = float(input("Ingrese el primer número: "))
    num2 = float(input("Ingrese el segundo número: "))
    return (num1 + num2) / 2 #se devuelve el número intermedio (el promedio de los dos números) 
#se llama a la función para calcular el número intermedio y se muestra el resultado
resultado = intermedio()
print(f"El número intermedio es: {resultado}")
```

**7-Función recortar**

```python
def recortar(numero, inferior, superior):#función para recortar un número a un rango definido por un límite inferior y un límite superior
    if numero < inferior:#si el número es menor que el límite inferior, se devuelve el límite inferior
        return inferior
    elif numero > superior:#si el número es mayor que el límite superior, se devuelve el límite superior
        return superior
    else:#si el número está dentro del rango, se devuelve el número original
        return numero

# Valores definidos directamente
numero = 15
limite_inferior = 0
limite_superior = 10

resultado = recortar(numero, limite_inferior, limite_superior)#se llama a la función para recortar el número y se muestra el resultado

print(f"El número recortado es: {float(resultado)}")
```

**8-Función separar()**


```python
def separar():#función para separar números pares e impares de una lista y ordenarlos
    numeros = [-12, 84, 13, 20, -33, 101, 9]
    pares = []
    impares = []
    for n in numeros:
        if n % 2 == 0:#si el número es par, se agrega a la lista de números pares
            pares.append(n)
        else:#si el número es impar, se agrega a la lista de números impares
            impares.append(n)
    #ordenamos las listas de números pares e impares
    pares.sort()
    impares.sort()
    return pares, impares
pares, impares = separar()#se llama a la función para separar los números pares e impares y se muestra el resultado
print(f"Números pares: {pares}")
print(f"Números impares: {impares}")
```

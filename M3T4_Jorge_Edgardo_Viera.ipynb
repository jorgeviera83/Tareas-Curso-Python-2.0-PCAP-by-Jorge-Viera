### 3- *Herencia simple en un sistema de viviendas*

```python
# ------------CODIGO CON HERENCIA-------------

class Vivienda:
    def __init__(self, direccion, metros_cuadrados: int, precio: float):
        self.direccion = direccion
        self.metros_cuadrados = metros_cuadrados
        self.precio = precio

    def __str__(self):
        return f"Vivienda en {self.direccion}, {self.metros_cuadrados} m², Precio: ${self.precio}"
    
class Casa(Vivienda):
    def __init__(self, direccion, metros_cuadrados: int, precio: float, jardin = bool):
        super().__init__(direccion, metros_cuadrados, precio)
        self.jardin = jardin

    def __str__(self):
        return f"Casa en {self.direccion}, {self.metros_cuadrados} m², Jardín: {self.jardin}, Precio: ${self.precio}"
    
class Apartamento(Vivienda):
    def __init__(self, direccion, metros_cuadrados: int, precio: float, piso: int):
        super().__init__(direccion, metros_cuadrados, precio)
        self.piso = piso

    def __str__(self):
        return f"Apartamento en {self.direccion}, {self.metros_cuadrados} m², Piso: {self.piso}, Precio: ${self.precio}"
    
#------ FASE DE EXPERIMENTACION -------

casa1 = Casa("Calle Roble 123", 120, 250000, jardin=True)
apartamento1 = Apartamento("Avenida Central 456, Piso 3", 80, 150000, piso=3)
print(casa1)
print(apartamento1)
# modificar atributos heredados
casa1.precio = 240000
apartamento1.precio = 140000
print(casa1)
print(apartamento1)
```

### 6- *Herencia múltiple y polimorfismo en un sistema de vehículos*

- Herencia limpia: Al dejar las clases hijas vacías con pass, heredan el comportamiento puro de los padres.

- Polimorfismo real: Invoco el mismo método (repostar()) en ambos objetos, pero actúan de forma totalmente distinta según el orden de sus padres en la declaración.

- **Tambien comprobamos el MRO, Method Resolution Order (Orden de Resolución de Métodos)**


```python
#------------CODIGO CON HERENCIA MÚLTIPLE Y MÉTODOS ESPECÍFICOS-------------

class Vehiculo:
    def repostar(self):
        print("Repostando gasolina...")

class Electrico:
    def repostar(self):
        print("Repostando electricidad...")

class BiciletaElectrica(Electrico, Vehiculo):
    pass  # No se define un método repostar específico, se usará el de Electrico debido al orden de herencia

class QuadHibrido(Vehiculo, Electrico):
    pass  # No se define un método repostar específico, se usará el de Vehiculo debido al orden de herencia (Vehiculo, Electrico)

#------ FASE DE EXPERIMENTACION -------

bicicleta = BiciletaElectrica()
quad = QuadHibrido()
bicicleta.repostar()  # Debería llamar al método de Electrico
quad.repostar()       # Debería llamar al método de Vehiculo debido al orden de herencia (Vehiculo, Electrico) en QuadHibrido

# comprobar MRO (Método de Resolución de Orden de Método) en Python para entender qué método se llama en cada caso.
print("MRO de BiciletaElectrica:")
for clase in BiciletaElectrica.__mro__:
    print(f" {clase.__name__}")

print("MRO de QuadHibrido:")
for clase in QuadHibrido.__mro__:
    print(f" {clase.__name__}")
```

### 7- *Herencia con personal universitario*

**Beneficios a tener en cuenta en el siguiente codigo**

-Seguridad de datos: Al usar datos=None, evitamos que diferentes instancias compartan accidentalmente la misma memoria del diccionario.

-Sintaxis de Diccionario: Al instanciar pasamos estructuras válidas de tipo {"clave": "valor"}.

-Legibilidad: La impresión en consola ahora está etiquetada y estructurada de forma más limpia.


```python
class Personal_Universitario:
    # Usamos None por defecto para evitar problemas de mutabilidad
    def __init__(self, datos=None):
        # Si no se pasan datos, se crea un diccionario con las claves básicas vacías
        if datos is None:
            self.datos = {"id": "", "nombre": "", "email": ""}
        else:
            self.datos = datos


class Oficina (Personal_Universitario):
    def __init__(self, datos = None, puesto = None):
        super().__init__(datos)
        self.puesto = puesto

class Profesor (Personal_Universitario):
    def __init__(self, datos = None, especializacion = None):
        super().__init__(datos)
        self.especializacion = especializacion

class Alumno (Personal_Universitario):
    def __init__(self, datos = None, creditosAprobados = None):
        super().__init__(datos)
        self.creditosAprobados = creditosAprobados

        #------ FASE DE EXPERIMENTACION -------

personal1 = Personal_Universitario( {"id": "PU001", "nombre": "Laura Martínez", "email": "laura.martinez@universidad.edu" })
oficina1 = Oficina( {"id": "OF001", "nombre": "Carlos Rodríguez", "email": "carlos.rodriguez@universidad.edu" }, puesto="Secretario" )
profesor1 = Profesor( {"id": "PR001", "nombre": "María González", "email": "maria.gonzalez@universidad.edu" }, especializacion="Matemáticas" )
alumno1 = Alumno( {"id": "AL001", "nombre": "Juan Pérez", "email": "juan.perez@universidad.edu" }, creditosAprobados=60 )
print(personal1.datos)
print(oficina1.datos, oficina1.puesto)
print(profesor1.datos, profesor1.especializacion)
print(alumno1.datos, alumno1.creditosAprobados)
```

### 8- *Herencia multinivel y polimorfismo (tema libre)*

*Es multinivel: Sigue la línea vertical* 
- Electrodomestico 
- DispositivoVisual 
- SmartTV.


*Tiene 5 clases:*
- Electrodomestico, 
- DispositivoLavado, 
- DispositivoVisual, 
- LavadoraInteligente
- SmartTV.
Demuestra polimorfismo: El bucle ejecuta .funcionar() en todos los objetos, y cada uno responde con un comportamiento adaptado a su clase.

```python
# 1. JERARQUÍA MULTINIVEL (5 clases)


# Nivel 1: Clase Base
class Electrodomestico:
    def __init__(self, marca):
        self.marca = marca  # Atributo que modificaremos

    def funcionar(self):
        return f"El electrodoméstico {self.marca} está encendido y consumiendo energía."

# Nivel 2: Clases Intermedias (Heredan de Electrodomestico)
class DispositivoLavado(Electrodomestico):
    def funcionar(self):
        return f"El dispositivo de lavado {self.marca} está preparando el suministro de agua."

class DispositivoVisual(Electrodomestico):
    def funcionar(self):
        return f"El dispositivo visual {self.marca} está encendiendo la pantalla."

# Nivel 3: Clases Específicas (Heredan de las intermedias)
class LavadoraInteligente(DispositivoLavado):
    def __init__(self, marca, revoluciones):
        super().__init__(marca)
        self.revoluciones = revoluciones

    # Sobrescritura para Polimorfismo
    def funcionar(self):
        return f"La lavadora {self.marca} está centrifugando a {self.revoluciones} RPM."

class SmartTV(DispositivoVisual):
    def __init__(self, marca, pulgadas):
        super().__init__(marca)
        self.pulgadas = pulgadas

    # Sobrescritura para Polimorfismo
    def funcionar(self):
        return f"La SmartTV {self.marca} de {self.pulgadas}\" está transmitiendo contenido en streaming."


# ----- FASE DE EXPERIMENTACION -------

print("--- Demostración de Polimorfismo ---")

# Creamos objetos de diferentes niveles de la jerarquía
dispositivos = [
    Electrodomestico("Genérico"),                       # Nivel 1
    DispositivoLavado("Samsung Base"),                  # Nivel 2
    LavadoraInteligente("LG", revoluciones=1200),       # Nivel 3
    SmartTV("Sony", pulgadas=55)                        # Nivel 3
]

# Todos responden al mismo método 'funcionar()' de manera diferente
for aparato in dispositivos:
    print(f"[{type(aparato).__name__}]: {aparato.funcionar()}")


print("\n--- Modificación de atributos y repetición ---")

# Modificamos los atributos de las clases específicas
dispositivos[2].marca = "LG Turbo"
dispositivos[2].revoluciones = 1400

dispositivos[3].marca = "Sony Bravia"
dispositivos[3].pulgadas = 65

# Repetimos la prueba con los objetos modificados
for aparato in dispositivos[2:]:
    print(f"[{type(aparato).__name__} Modificado]: {aparato.funcionar()}")
```

## 1-Listas y Tuplas

- Lista y Tupla

```python

#Lista principal para tarea
lista_numeros_pares = [2, 4, 6, 8, 10] 

#Anexo de tipos de listas
lista_vacia = []
lista_nombres = ["Jorge", "Natalia", "Homero"]
lista_mixta = [12, "Homero", 6.5, True]

print(lista_vacia)
print(lista_numeros_pares)
print(lista_nombres)
print(lista_mixta)


#Tupla principal para tarea
tupla_numeros_pares = (2, 4, 6, 8, 10) 

#Anexo de tipos de tuplas
tupla_vacia = ()
tupla_nombres = ("Jorge", "Natalia", "Homero")
tupla_mixta = (12, "Homero", 6.5, True)


tupla_de_un_elemento = (26,) # Si la Tupla es de un solo elemento (es necesario poner la coma)

print(tupla_vacia)
print(tupla_numeros_pares)
print(tupla_nombres)
print(tupla_mixta)
print(tupla_de_un_elemento)
```

**Muestra el segundo elemento de la lista y el penúltimo de la tupla.**

```python
# Acceso por índice a los elementos de la lista
print(lista_numeros_pares[1])
# Acceso por índice a los elementos de la tupla
print(tupla_numeros_pares[-2])
```

**Modifica (si se puede) algún elemento de la lista y de la tupla y muestra el resultado.**

```python
# Modificación de un elemento de la lista
lista_numeros_pares[1] = 12 
print(lista_numeros_pares)

# Intento de modificación de un elemento de la tupla (esto generará un error) las tuplas son inmutables, no se pueden modificar después de su creación
tupla_numeros_pares[1] = 12
print(tupla_numeros_pares)
```

**Muestra el tamaño de la lista y de la tupla**

```python
# Imprime la longitud de la lista
print(len(lista_numeros_pares))

# Imprime la longitud de la tupla
print(len(tupla_numeros_pares)) 
```

**Búsqueda de un elemento con el operador in**

```python
# Verifica si el número 26 está en la lista
print(12 in lista_numeros_pares) 

# Verifica si el número 26 está en la tupla
print(26 in tupla_numeros_pares) 
```

**Añadir un elemento a una Lista y a una Tupla**

```python
# Agregar un elemento al final de la lista
lista_numeros_pares.append(55) 
print(lista_numeros_pares)

# Agregar un elemento a la tupla (esto no es posible, las tuplas son inmutables)
tupla_numeros_pares.append(55) #descomentar esta línea para evitar el error
print(tupla_numeros_pares)
```

**Borrar o eliminar un elemento de una Lista y una Tupla**

```python
#elimina el elemento 55 de la lista
lista_numeros_pares.remove(55)
print(lista_numeros_pares)

# Eliminar por posición y devolver el elemento eliminado
elemento_eliminado = lista_numeros_pares.pop()
print(elemento_eliminado)
print(lista_numeros_pares)

tupla_de_numeros_pares.remove(55) 
print(tupla_de_numeros_pares)
```

## 2-Sets y Diccionarios

**Set (Conjunto)**

```python
set() #se define un conjunto vacío

#se define con {} si hay elementos
conjunto_nombres = {"Jorge", "Natalia", "Homero"} 
print(conjunto_nombres)

#muestra el segundo elemento del conjunto (esto no es posible, los conjuntos no tienen orden)
print(conjunto_nombres[1]) #descomentar esta línea para evitar el error

#modifica un elemento del conjunto (esto no es posible, los conjuntos no tienen orden ni índices)
conjunto_nombres[1] = "Juan" #descomentar esta línea para evitar el error
```

**Dict (Diccionario)**

```python
dic_vacio = {} #diccionario vacío
diccionario_persona = {"nombre": "Jorge", "edad": 30, "ciudad": "Buenos Aires"}
print(dic_vacio)
print(diccionario_persona)

#muestra el valor asociado a la clave "nombre"
print(diccionario_persona["nombre"])

#editar valor asociado a la clave "edad"
diccionario_persona["edad"] = 31
print(diccionario_persona["edad"])
```

 **Tamaño de ambos con len()**

```python
 #muestra la cantidad de elementos en el conjunto
len(conjunto_nombres)

 #muestra la cantidad de pares clave-valor en el diccionario
len(diccionario_persona) 
```

**Búsqueda en ambos (con in para set y in/.get() para diccionario).**

```python
#pertenece un elemento al conjunto
print("Jorge" in conjunto_nombres)

#pertenece una clave al diccionario
print("nombre" in diccionario_persona)

print(diccionario_persona.get("ciudad"))

#anadir un nuevo elemento al conjunto
conjunto_nombres.add("Juan")
print(conjunto_nombres)

#anadir un nuevo par clave-valor al diccionario
diccionario_persona["Hijo"] = "Juani"
print(diccionario_persona)

#borrar el contenido del conjunto
conjunto_nombres.clear()

#borrar el contenido del diccionario
diccionario_persona.clear()
```

## 5-Lista de la compra

```python
lista_de_compras = ["leche", "pan", "huevos", "frutas", "verduras", "jugo",]
print(lista_de_compras)

#mostrar el segundo elemento de la lista de compras
print(lista_de_compras[1])
print(lista_de_compras[-5])

#mostrar el penúltimo elemento de la lista de compras
print(lista_de_compras[-1])
print(lista_de_compras[5])

lista_de_compras[2] = "cereal"
print(lista_de_compras)

#intercambiar el primer y el último elemento de la lista de compras
lista_de_compras[0], lista_de_compras[-1] = lista_de_compras[-1], lista_de_compras[0]
print(lista_de_compras)

#Extrae con slicing los primeros tres elementos de la lista de compras
primeros_tres = lista_de_compras[:3]
print(primeros_tres)

#Extrae con slicing los últimos tres elementos de la lista de compras
ultimos_tres = lista_de_compras[-3:]
print(ultimos_tres)

#Extrae con slicing pocisiones pares de la lista de compras
posiciones_pares = lista_de_compras[::2]
print(posiciones_pares)
```

## 6-Formateo de cadenas invertidas

```python
#cadena corrupta de prueba
cadena = "zereP solraC, 5891"
#Extrae con slicing los caracteres de la cadena en orden inverso
cadena_inversa = cadena[::-1]
print(cadena_inversa)

#separar la cadena en palabras utilizando el espacio como separador
palabras = cadena_inversa.split()
print(palabras)

#dar formato a la cadena para que tenga la forma "Carlos Pérez, 1985"
nombre = palabras[1] + " " + palabras[2] + ", " + palabras[0]
print(nombre)


```

## 7-Operaciones con conjuntos

```python
mamiferos = {"Gato", "León", "Caballo", "Ballena", "Tortuga"}
print(mamiferos)
peces = {"Ballena", "Sardina"}
print(peces)

#borrar el elemento "Tortuga" del conjunto de mamíferos
mamiferos.remove("Tortuga")
print(mamiferos)

#Añadir el elemento "Pez Payaso" al conjunto de peces
peces.add("Pez Payaso")
print(peces)

#intersección de los conjuntos de mamíferos y peces
interseccion = mamiferos.intersection(peces)
print(interseccion)

```

## 8-Comparación de alumnos

```python
alumnos_Python = {"Jorge", "Natalia", "Homero", "Lisa"}
alumnos_IA = {"Jorge", "Natalia", "Homero", "Bart"}

#Alumnos que estan en ambos cursos (intersección de los conjuntos de alumnos de Python e IA)
interseccion = alumnos_Python.intersection(alumnos_IA)
print(interseccion)

#conversion de la intersección a una lista
interseccion_lista = list(interseccion)
print(interseccion_lista)


#alunos exclusivos de python (diferencia de los conjuntos de alumnos de Python e IA)
exclusivos_python = alumnos_Python.difference(alumnos_IA)
print(exclusivos_python)

#conversión de la diferencia a una lista
exclusivos_python_lista = list(exclusivos_python)
print(exclusivos_python_lista)


#alunos exclusivos de IA (diferencia de los conjuntos de alumnos de IA y Python)
exclusivos_IA = alumnos_IA.difference(alumnos_Python)
print(exclusivos_IA)

#conversión de la diferencia a una lista
exclusivos_IA_lista = list(exclusivos_IA)
print(exclusivos_IA_lista)

```

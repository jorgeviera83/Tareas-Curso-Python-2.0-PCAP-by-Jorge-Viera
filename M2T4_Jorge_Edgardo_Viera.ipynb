### Primer bloque — División por cero

*En Python, al igual que en la mayoría de los lenguajes de programación y en las matemáticas convencionales, la división por cero no está definida y el intérprete detiene la ejecución del programa inmediatamente al encontrar esta instrucción.*

**El código proporcionado provocará un error de ejecución de tipo (ZeroDivisionError).**


```python
# Código a evaluar con solución (try except) implementada
try:
    numero = 7/0 # Intento de división por cero de un número provoca una excepción ZeroDivisionError
except ZeroDivisionError:
    print("Error: División por cero no permitida.")
```

### Segundo bloque — Índice fuera de rango

*Esto sucede porque la lista solo tiene 5 elementos (índices del 0 al 4). Al intentar acceder a lista[10], el programa busca una posición que no existe en la memoria asignada a esa variable.*

**El código arrojará un error de tipo IndexError: list index out of range.**

```python
# Código a evaluar con solución (try except) implementada
lista = [4, 7, 30, 23, 5]
try:
    lista[10] # Intento de acceder a un índice fuera del rango de la lista provoca una excepción IndexError
except IndexError:
    print("Error: Índice fuera de rango.")
```

### Tercer bloque — Clave inexistente en diccionario

*Esto sucede cuando se intenta acceder a la clave "alemania", pero esta no existe dentro del diccionario paises. En Python, si buscas una clave inexistente directamente con corchetes, el programa se detiene y lanza KeyError*

**El código provocará un error de tipo (KeyError).**



```python
# Código a evaluar con solución (try except) implementada
paises = {"españa": "español", "eeuu": "inglés", "italia": "italiano"}
try:
    paises["alemania"] # Intento de acceder a una clave que no existe en el diccionario provoca una excepción KeyError
except KeyError:
    print("Error: Clave no encontrada.")
```

### Cuarto bloque — Error de tipo

*En la mayoría de los lenguajes de programación, como Python, no se puede sumar directamente un texto ("2") con un número (10) porque son tipos de datos diferentes.*

**El código provocará un error de tipo (TypeError).**

```python
# Código a evaluar con solución (try except) implementada
try:
    Resultado = "2" + 10 # Intento de concatenar una cadena con un número provoca una excepción TypeError
except TypeError:
    print("Error: No se puede concatenar una cadena con un número.")
```

### Quinto bloque — Claves de entrada de usuario

*Este bloque de código muestra cómo acceder a un valor dentro de un diccionario usando una entrada dinámica del usuario. Sin embargo, tiene un detalle importante a tener en cuenta:
Este código es frágil. Si el usuario introduce una clave que no existe (por ejemplo, "apellido" o comete una falta de ortografía), el programa se detendrá* **inmediatamente con un error de tipo KeyError.**

```python
# Código a evaluar con solución (try except) implementada
datos = {
  "nombre": "Juan",
  "edad": 30,
  "ciudad": "Madrid"
}
clave = input("Introduce una clave para obtener su valor: ")
try:
    print(f"El valor asociado a '{clave}' es: {datos[clave]}")
except KeyError: # Intento de acceder a una clave que no existe en el diccionario provoca una excepción KeyError
    print("Error: Clave no encontrada.")
```

### Sexto bloque — Conversión y división

*En el código original proporcionado pueden ocurrir principalmente dos excepciones:

1-ValueError: Se activará si el usuario introduce algo que no sea un número entero (por ejemplo, letras o símbolos). Esto sucede porque la función int() no podrá convertir la cadena de texto a un valor numérico.*


*2-ZeroDivisionError: Ocurrirá si el usuario introduce un 0 como segundo número. Aunque la conversión a entero sea exitosa, la división final fallará porque no es posible dividir por cero.*

```python
# Código a evaluar original:
def calcular(numero1, numero2):
    return (int(numero1) ** 2) / int(numero2)

n1 = input("Introduce el primer número: ")
n2 = input("Introduce el segundo número: ")
resultado = calcular(n1, n2)
print(f"El resultado del cálculo es: {resultado}")



# Código a evaluar con solución (try except) implementada
def calcular(numero1, numero2):
    return (int(numero1) ** 2) / int(numero2)

n1 = input("Introduce el primer número: ")
n2 = input("Introduce el segundo número: ")
try:
    resultado = calcular(n1, n2)
    print(f"El resultado del cálculo es: {resultado}")
except (ValueError, ZeroDivisionError) as e: # Intento de convertir una entrada no numérica a entero provoca una excepción ValueError, y división por cero provoca una excepción ZeroDivisionError
    print(f"Error: {e}")
```

### Séptimo bloque — Función para agregar sin repetir

*Se utiliza ValueError porque es la excepción estándar en Python cuando una función recibe un argumento que tiene el tipo correcto (por ejemplo, un número) pero un valor inapropiado (un duplicado, en este caso)*

**IMPORTANTE**: Utilixzamos *raise* al detectar el duplicado, el código detiene su ejecución normal. Es como poner un "freno de mano" antes de que la lista se ensucie con datos repetidos.

El uso de **if** elemento in lista obliga a Python a recorrer la lista elemento por elemento. Si la lista tuviera millones de datos, el "error" tardaría más en detectarse. Para casos masivos, se usarían estructuras como **set**.

```python
def agregar_sin_repetidos(lista, elemento):
    try:
        if elemento in lista:
            raise ValueError(f"Error: Imposible añadir elementos duplicados => {elemento}") # Si el elemento ya existe en la lista, se lanza una excepción ValueError con un mensaje personalizado
        lista.append(elemento)
        print(f"Elemento '{elemento}' agregado a la lista.")
    except ValueError as e: # Captura la excepción ValueError y muestra el mensaje de error personalizado
        print(e)
mi_lista = [1, 2, 3]
agregar_sin_repetidos(mi_lista, 4) # Agrega el elemento 4 a la lista
agregar_sin_repetidos(mi_lista, 2) # Intenta agregar el elemento 2 nuevamente, lo que provoca una excepción ValueError  
print(mi_lista) # Imprime la lista actualizada
```

### Octavo bloque — Investigando excepciones

**Explicación:**

-*AttributeError*: Se lanza porque reverse() es un método de las listas, no de los textos.

-*as e*: Captura la descripción técnica del sistema.

-*Mensaje personalizado*: Traduce el problema técnico a algo que el usuario pueda entender fácilmente. 

```python
try:
    nombre = "Mundo"
    # Las cadenas no tienen el método 'reverse'
    nombre.reverse() 
except AttributeError as e:
    print(f"Información por defecto: {e}")
    print("Mensaje personalizado: Estás intentando usar una función que no pertenece a este tipo de dato.")
```

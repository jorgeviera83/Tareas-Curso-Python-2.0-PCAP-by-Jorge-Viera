3. Formateo de números

```python
numero = 43
print(f"{43:05}") #imprime el numero 43 con ceros a la izquierda para que ocupe un total de 5 caracteres, dando como resultado "00043"

numero2 = 31415.9265

print(f"{numero2:.3f}") #imprime el numero 31415.9265 redondeado a 3 decimales, dando como resultado "31415.927"

```

4. Formas de formatear

* Distintas variantes de **format()**

```python
altura = 1.80
peso = 80.135

# ------ Usando f-strings (recomendado) ------
print(f"La altura es de {altura:.2f} metros y el peso es de {peso:.2f} kg.") # Imprime: La altura es de 1.80 metros y el peso es de 80.14 kg.

# ------ Usando format() ------
print("La altura es de {:.2f} metros y el peso es de {:.2f} kg.".format(altura, peso)) # Imprime: La altura es de 1.80 metros y el peso es de 80.14 kg. 

print("La altura es de {0} metros y el peso es de {1} kg.".format(altura, peso)) # Imprime: La altura es de 1.8 metros y el peso es de 80.135 kg. (sin formato específico)

```

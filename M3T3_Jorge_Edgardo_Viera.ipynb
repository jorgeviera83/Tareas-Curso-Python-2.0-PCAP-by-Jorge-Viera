- ### *Clase Alumno*

- Constructor `__init__`: Inicializa el nombre y la nota. Lanza el mensaje automático de éxito al instanciar el objeto.
- Método `__str__`: Permite que, al hacer **print(alumno1)**, se muestre un texto limpio y legible en lugar de la dirección de memoria del objeto.
- Método calificacion: Utiliza una estructura condicional básica para evaluar si la nota es igual o mayor a 6 (aprobado) o menor (suspendido).


```python
class Alumno:
    def __init__(self, nombre, nota):
        self.nombre = nombre
        self.nota = nota
        print(f"Alumno {self.nombre} creado con nota {self.nota}")

    def __str__(self):
        return f"Alumno: {self.nombre}, Nota: {self.nota}"
    
    def calificacion(self):
        if self.nota >= 6:
            return "Aprobado"
        else:
            return "Suspendido"
        
    # --- Fase de Experimentación ---
      
alumno1 = Alumno("Juan", 5)
print("print(alumno1):", alumno1)
print("Calificación:", alumno1.calificacion())

alumno2 = Alumno("María", 8)
print("print(alumno2):", alumno2)
print("Calificación:", alumno2.calificacion())

alumno3 = Alumno("Pedro", 6)
print("print(alumno3):", alumno3)
print("Calificación:", alumno3.calificacion())
```

- ### *Encapsulación de la clase Alumno*

Se modifica el ejercicio anterior para encapsular los atributos de la clase con `@property` y se crean los metodos getter y setter

**NOTA IMPORTANTE**: Como en Python no existen los atributos 100% privados. Por convención, usamos un guion bajo (_nombre) para indicar que es privado

- El print nos informa a modo de prueba donde enviamos la invormacion. Esto demuestra, mostrandolo por consola, que Python no cambió la variable directamente, sino que invocó tu método de control.
- La validación protege el atributo: Al intentar poner un 15, el setter actúa como un escudo y lanza un error antes de modificar el valor real (_nota).


```python
class Alumno:
    def __init__(self, nombre, nota):
        self._nombre = nombre # el guion bajo es una convención para indicar que es un atributo privado
        self._nota = nota
        print(f"Alumno {self._nombre} creado con nota {self._nota}")

    @property
    def nombre(self):
        print("Accediendo al nombre del alumno en el getter") # el print es para mostrar que se está accediendo al getter
        return self._nombre
        
    @nombre.setter
    def nombre(self, valor):
        print("Modificando el nombre del alumno en el setter") # el print es para mostrar que se está accediendo al setter
        self._nombre = valor

    @property
    def nota(self):
        print("Accediendo a la nota del alumno en el getter")
        return self._nota
        
    @nota.setter
    def nota(self, valor):
        print("Modificando la nota del alumno en el setter")
        if valor < 0 or valor > 10:
            raise ValueError("La nota debe estar entre 0 y 10")
        self._nota = valor

  
    def __str__(self):
        return f"Alumno: {self.nombre}, Nota: {self.nota}"
    
    def calificacion(self):
        if self.nota >= 6:
            return "Aprobado"
        else:
            return "Suspendido"
        

alumno1 = Alumno(nombre="Juan", nota=6)
print(f"{alumno1.nombre} tiene nota {alumno1.nota} y está {alumno1.calificacion()}")
alumno2 = Alumno(nombre="María", nota=8)
print(f"{alumno2.nombre} tiene nota {alumno2.nota} y está {alumno2.calificacion()}")

# Esto activará el setter de forma automática
alumno1.nota = 9 

print("\n--- Intentar saltarse las reglas poniendo una nota inválida ---")
try:
    alumno1.nota = 15  # El setter bloqueará esto
except ValueError as e:
    print(f"Bloqueado por el setter: {e}")

```

- ### *Clase Producto*

- Usar @property permite seguir leyendo y modificando los datos como si fueran atributos normales, pero por detrás se están ejecutando tus métodos seguros.
- *Validación interna*: He añadido una regla en el precio.setter para evitar precios negativos.

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo):
        self._codigo = codigo
        self._nombre = nombre
        self._precio = precio
        self._tipo = tipo
        print(f"Producto {self._nombre} creado con código {self._codigo}, precio {self._precio}€ y tipo {self._tipo} exitosamente")
    
    @property
    def codigo(self):
        return self._codigo
    
    @codigo.setter
    def codigo(self, valor):
        self._codigo = valor

    @property
    def nombre(self):
        return self._nombre
    
    @nombre.setter
    def nombre(self, valor):
        self._nombre = valor

    @property
    def precio(self):
        return self._precio

    @precio.setter
    def precio(self, valor):
        if valor < 0:
            raise ValueError("El precio no puede ser negativo")
        self._precio = valor

    @property
    def tipo(self):
        return self._tipo
    
    @tipo.setter
    def tipo(self, valor):
        self._tipo = valor
        
# Fase de Experimentación
producto1 = Producto(codigo="P001", nombre="Camiseta", precio=19.99, tipo="Ropa")
producto2 = Producto(codigo="P002", nombre="Zapatos", precio=49.99, tipo="Calzado")

# lectura con getters
print(f"Producto 1: Código={producto1.codigo}, Nombre={producto1.nombre}")
print(f"Producto 2: Nombre={producto2.nombre}, Precio={producto2.precio}")

# Modificación con setters
producto1.precio = 17.99 # Esto activará el setter y actualizará el precio
print(f"Producto 1 actualizado: Código={producto1.codigo}, Nombre={producto1.nombre}, Precio={producto1.precio}€")
producto2.tipo = "Accesorios" # Esto activará el setter y actualizará el tipo
print(f"Producto 2 actualizado: Código={producto2.codigo}, Nombre={producto2.nombre}, Tipo={producto2.tipo}")

```

- ### *Ampliación de la clase Producto*

Sobrescribimos el método `__str__`() para que, al imprimir un objeto Producto, se muestren todos sus datos de forma estructurada.

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo):
        self._codigo = codigo
        self._nombre = nombre
        self._precio = precio
        self._tipo = tipo
        print(f"Producto {self._nombre} creado con código {self._codigo}, precio {self._precio}€ y tipo {self._tipo} exitosamente")
    
    @property
    def codigo(self):
        return self._codigo
    
    @codigo.setter
    def codigo(self, valor):
        self._codigo = valor

    @property
    def nombre(self):
        return self._nombre
    
    @nombre.setter
    def nombre(self, valor):
        self._nombre = valor

    @property
    def precio(self):
        return self._precio

    @precio.setter
    def precio(self, valor):
        if valor < 0:
            raise ValueError("El precio no puede ser negativo")
        self._precio = valor

    @property
    def tipo(self):
        return self._tipo
    
    @tipo.setter
    def tipo(self, valor):
        self._tipo = valor
    
    def __str__(self):
        return (
            f"Código: {self.codigo}\n"
            f"Nombre: {self.nombre}\n"
            f"Precio: {self.precio}€\n"
            f"Tipo: {self.tipo}"
        )
    # Fase de Experimentación
    def calcular_total(self, cantidad):
        if cantidad < 0:
            raise ValueError("La cantidad no puede ser negativa")
        return self.precio * cantidad
    

producto1 = Producto(codigo="1", nombre="Macbook Pro", precio=1000, tipo="Informática")
#  al imprimir un objeto Producto, se muestren todos sus datos de forma estructurada.
print(producto1)
producto2 = Producto(codigo="2", nombre="Camiseta", precio=25, tipo="Ropa")
print(producto2)

print(producto1.calcular_total(5)) # Esto activará el método calcular_total y mostrará el total para 5 unidades
```

- ### *Clase Empleado*

```python
class Empleado:
    def __init__(self, id,nombre, puesto, salario):
        self._id = id
        self._nombre = nombre
        self._puesto = puesto
        self._salario = salario
        print(f"Empleado {self._nombre} creado con id {self._id}, puesto {self._puesto} y salario {self._salario:.2f}€ exitosamente") 
    @property
    def id(self):
        return self._id
    
    @id.setter
    def id(self, valor):
        self._id = valor

    @property
    def nombre(self):
        return self._nombre
    
    @nombre.setter
    def nombre(self, valor):
        self._nombre = valor

    @property
    def puesto(self):
        return self._puesto 
    
    @puesto.setter
    def puesto(self, valor):
        self._puesto = valor

    @property
    def salario(self):
        return self._salario

    @salario.setter
    def salario(self, valor):
        if valor < 0:
            raise ValueError("El salario no puede ser negativo")
        self._salario = valor
    
    def __str__(self):
        return f"Empleado: {self.nombre}, Salario: {self.salario:.2f}€"
    def mostrar_informacion(self):
        return (
            f"ID: {self.id}\n"
            f"Nombre: {self.nombre}\n"
            f"Puesto: {self.puesto}\n"
            f"Salario: {self.salario}€"
        )
    def incrementar_salario(self, porcentaje):
        if porcentaje < 0:
            raise ValueError("El porcentaje no puede ser negativo")
        self.salario += self.salario * (porcentaje / 100)
# Fase de Experimentación
empleado1 = Empleado(id=1, nombre="Ana", puesto="Desarrolladora", salario=3000)
print(empleado1.mostrar_informacion())

empleado2 = Empleado(id=2, nombre="Luis", puesto="Diseñador", salario=2500)
print(empleado2.mostrar_informacion())

# modificar el puesto de Ana a "Senior Desarrolladora" usando el setter
empleado1.puesto = "Senior Desarrolladora" # Esto activará el setter y actualizará el puesto
print("\nInformación actualizada del empleado:")
print(empleado1.mostrar_informacion())

# Incrementar el salario de Ana en un 10%
empleado1.incrementar_salario(10) # Esto activará el método incrementar_salario y aumentará el salario en un 10%
print("\nInformación después del incremento salarial:")
print(empleado1.mostrar_informacion())
```

### 1- Transformaciones básicas con map y lambda

**Acciones en el siguiente codigo**

- Aplicamos map y lambda correctamente para las tres transformaciones.
- Generamos las cadenas de texto de forma dinámica desde la lista original.
- Validamos las longitudes de forma limpia.

```python
#------------ FASE DE CREACION -----------

numeros_enteros = [1, 2, 3, 4, 5]
lista_de_cuadrados = lambda x: x ** 2 # Función lambda para calcular el cuadrado de un número
cuadrados = list(map(lista_de_cuadrados, numeros_enteros)) # Aplicamos la función a cada elemento de la lista
print(cuadrados)

# calculamos los dobles de los números enteros utilizando una función lambda
dobles = list(map(lambda x: x * 2, numeros_enteros)) # Función lambda para calcular el doble de un número
print(dobles)

# una lista de cadenas con el formato N=<valor>
cadenas = list(map(lambda x: f"N={x}", numeros_enteros))
print(cadenas)

#----------- FASE DE EXPERIMENTACION -----------
print("\n------------ FASE DE EXPERIMENTACION -----------")
#mostrar las tres listas resultantes
print("Lista de cuadrados:", cuadrados)
print("Lista de dobles:", dobles)
print("Lista de cadenas:", cadenas)

#comprobar que las tres listas tienen la misma longitud con respecto a la lista original de números enteros
if len(cuadrados) == len(numeros_enteros) and len(dobles) == len(numeros_enteros) and len(cadenas) == len(numeros_enteros):
    print("Las tres listas tienen la misma longitud que la lista original de números enteros.")
else:    print("Las tres listas no tienen la misma longitud que la lista original de números enteros.")

# transformamos con map la lista de cadenas para para agregar el sufijo "ok"
cadenas_con_ok = list(map(lambda s: s + "ok", cadenas))
print("Lista de cadenas con 'ok':", cadenas_con_ok)
```

### 2- Filtrado con filter y selección con list comprehensions

Hemos logrado en este codigo:
- Mostrar la lista original, la de aprobados y la de nombres. **mediante la creacion de la lista de dcit y la utilizacion de **`lambda`, `filter` y `list comprehension`**
- Verificar que ningún alumno con nota menor de 5 aparezca entre los aprobados utilizando `all()`, `map()` y `lambda`.
- Y la tarea Opcional que nos solicitaba indicar cómo ordenaría la lista de aprobados por nota descendente.

```python
#----------- FASE DE CREACION -----------
#creamos una lista de alumnos como diccionarios con nombre y nota
alumnos = [
    {"nombre": "Miguel", "nota": 4.5},
    {"nombre": "Douglas", "nota": 6.0},
    {"nombre": "Jorge", "nota": 8.75}
]

#usamos filter y lambda para obtener una lista de alumnos aprobados (nota >= 5.0)
alumnos_aprobados = list(filter(lambda alumno: alumno["nota"] >= 5.0, alumnos))
print("Alumnos aprobados:", alumnos_aprobados)

# Con una list comprehension, podemos extraer la lista de nombres de los aprobados.
nombres_aprobados = [alumno["nombre"] for alumno in alumnos_aprobados]
print("Nombres de alumnos aprobados:", nombres_aprobados)

#Con otra list comprehension, creamos una lista con f"{nombre} - {nota_redondeada}" (nota redondeada a 1 decimal)
alumnos_con_nota = [f"{alumno['nombre']} - {alumno['nota']:.1f}" for alumno in alumnos]
print("Alumnos con notas redondeadas:", alumnos_con_nota)

# ----------- FASE DE EXPERIMENTACION -----------
# Muestra la lista original, la de aprobados y la de nombres.
print("\n------------ FASE DE EXPERIMENTACION -----------")
print("Lista original:", alumnos)
print("Alumnos aprobados:", alumnos_aprobados)
print("Nombres de alumnos aprobados:", nombres_aprobados)

# Verificamos que ningún alumno con nota menor de 5 aparezca entre los aprobados.
verificacion = all(map(lambda alumno: alumno["nota"] >= 5.0, alumnos_aprobados))

if verificacion:
    print("Verificación correcta: Todos los alumnos aprobados tienen nota >= 5.0")
else:
    print("Error: Se coló un alumno con nota menor a 5.0")

# Ordenamos la lista de aprobados por nota descendente y mostramos el resultado.
alumnos_aprobados_ordenados = sorted(alumnos_aprobados, key=lambda alumno: alumno["nota"], reverse=True)
print("Alumnos aprobados ordenados por nota descendente:", alumnos_aprobados_ordenados)
```

### 3- Comprehensions mixtas: listas, conjuntos y diccionarios

*Aplicando de forma correcta las tres estructuras solicitadas: `list comprehension`, `set comprehension` y `dict comprehension`*

```python
#----------- FASE DE CREACION -----------


emails = ["ana@empresa.com", "luis@uni.es", "ana@empresa.com", "marta@empresa.com"]

# Con list comprehension extraemos una lista de dominios lo que va despues de la @
dominios = [email.split("@")[1] for email in emails]
print("Dominios:", dominios)

# Con set comprehension obtenemos un conjunto de dominios únicos
dominios_unicos = {email.split("@")[1] for email in emails}
print("Dominios únicos:", dominios_unicos)

#Con una dict comprehension, crea un diccionario {dominio: cantidad_apariciones} contando cuántas veces aparece cada dominio
conteo_dominios = {dominio: dominios.count(dominio) for dominio in dominios_unicos}
print("Conteo de dominios:", conteo_dominios)

#----------- FASE DE EXPERIMENTACION -----------

# Muestra la lista de dominios, el conjunto (sin duplicados) y el diccionario de frecuencias.
print("\n------------ FASE DE EXPERIMENTACION -----------")
print("Lista de dominios:", dominios)
print("Conjunto de dominios únicos:", dominios_unicos)
print("Diccionario de conteo de dominios:", conteo_dominios)

# Creamos un diccionario que asigne a cada email su longitud.
longitudes_emails = {email: len(email) for email in emails}
print("Longitudes de emails:", longitudes_emails)

# Creamos una lista que filtre solo los emails con dominio "empresa.com"
emails_empresa = [email for email in emails if email.split("@")[1] == "empresa.com"]
print("Emails con dominio 'empresa.com':", emails_empresa)
```

### 4- Pipeline funcional con filter + map y alternativa con list comprehensions

*EN ESTE EJERCICIO:*

- Logramos **Inmutabilidad** calculando el descuento directamente al construir las tuplas tanto en el `map` como en la `list comprehension`. Esto evitó modificar los diccionarios originales de la lista productos, garantizando que la fase de experimentación sea limpia y exacta.

- El uso de precio `for _, precio in precio_descuento` es la mejor forma de ignorar el nombre del producto elegantemente *usando el guion bajo _* y solo sumar los valores numericos.


```python
#----------- FASE DE CREACION -----------
# Creamos una lista de productos (diccionarios) con nombre, precio y stock. uno tiene stock > 0 y otro precio ≥ 10
productos = [
    {"nombre": "Producto A", "precio": 15.0, "stock": 5},
    {"nombre": "Producto B", "precio": 8.0, "stock": 0},
    {"nombre": "Producto C", "precio": 12.0, "stock": 3}
]

# Usando filter + lambda, selecciona solo los productos con stock > 0 y precio ≥ 10
productos_filtrados = list(filter(lambda p: p["stock"] > 0 and p["precio"] >= 10, productos))
print("Productos filtrados:", productos_filtrados)

# Usando map + lambda, aplica un 10% de descuento al precio de los filtrados y genera tuplas (nombre, precio_descuento).
precio_descuento = list(map(lambda p: (p["nombre"], p["precio"] * 0.9), productos_filtrados))
print("Precios con descuento:", precio_descuento)

# List comprehension, para crear una alternativa equivalente en una sola expresión que devuelva las mismas tuplas.
precio_descuento_lc = [(p["nombre"], p["precio"] * 0.9) for p in productos if p["stock"] > 0 and p["precio"] >= 10]
print("Precios con descuento (list comprehension):", precio_descuento_lc)

#----------- FASE DE EXPERIMENTACION -----------
print("\n------------ FASE DE EXPERIMENTACION -----------")
# Muestra la lista original de productos, la filtrada, y las tuplas con descuento.
print("Lista original de productos:", productos)
print("Productos filtrados:", productos_filtrados)
print("Precios con descuento:", precio_descuento)
print("Precios con descuento (list comprehension):", precio_descuento_lc)

# Verificamos que ambas salidas con descuento sean iguales.
if precio_descuento == precio_descuento_lc:
    print("Verificación correcta: Ambas listas de precios con descuento son iguales.")
else:    print("Error: Las listas de precios con descuento no son iguales.")

# Calculamos el total de la compra sumando los precios con descuento.
total_compra = sum(precio for _, precio in precio_descuento) # desempaquetado limpiamente con _ para el nombre
print("Total de la compra con descuento:", total_compra)

```

### 5- Gestión de encuestas: limpieza y análisis funcional

```python
sedeA = [5, 4, 3, 5]
sedeB = [2, 3, 3, 4]
sedeC = [1, 5, 5, 2]

# Usamos una list comprehesion anidada para aplanar todos los resultados en una sola lista
todas_las_sedes = [nota for sede in [sedeA, sedeB, sedeC] for nota in sede]
print("Todas las notas de las sedes:", todas_las_sedes)

# set comprehension para identificar las puntuaciones únicas recogidas (valores de 1 a 5).
puntuaciones_unicas = {nota for sede in [sedeA, sedeB, sedeC] for nota in sede}
print("Puntuaciones únicas:", puntuaciones_unicas)

# usando filter + lambda, solo las respuestas altas (≥ 4).
respuestas_altas = list(filter(lambda nota: nota >= 4, todas_las_sedes))
print("Respuestas altas (≥ 4):", respuestas_altas)

# usamos map + lambda y transformamos las respuestas en cadenas del tipo "Excelente (5)" o "Buena (4)"
transformacion_altas = list(map(lambda nota: f"Excelente ({nota})" if nota == 5 else f"Buena ({nota})", respuestas_altas))
print("Transformación de respuestas altas:", transformacion_altas)

# Ahora cambiamos el criterio de filtrado para anlizar respuestas bajas (≤ 2) sin modificar la estructura funcional
respuestas_bajas = list(filter(lambda nota: nota <= 2, todas_las_sedes))
print("Respuestas bajas (≤ 2):", respuestas_bajas)

transformacion_bajas = list(map(lambda nota: f"Regular ({nota})" if nota == 2 else f"Mala ({nota})", respuestas_bajas))
print("Transformación de respuestas bajas:", transformacion_bajas)

#----------- FASE DE EXPERIMENTACION -----------
print("\n------------ FASE DE EXPERIMENTACION -----------")
#Muestra la lista original (matriz de respuestas), la lista aplanada, el set de puntuaciones únicas, los filtrados altos y su transformación textual.
print("Matriz de respuestas (sedes):", [sedeA, sedeB, sedeC])
print("Todas las notas de las sedes (aplanada):", todas_las_sedes)
print("Puntuaciones únicas:", puntuaciones_unicas)

print("\n--- Bloque Respuestas Altas ---")
print("Respuestas altas (≥ 4):", respuestas_altas)
print("Transformación de respuestas altas:", transformacion_altas)

print("\n--- Bloque Respuestas Bajas ---")
print("Respuestas bajas (≤ 2):", respuestas_bajas)
print("Transformación de respuestas bajas:", transformacion_bajas)
```
